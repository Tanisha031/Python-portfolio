# **Markdown to HTML Converter**

```python
# Step 1: Import required libraries
import markdown
from IPython.display import display, HTML
```

```python
# Step 2: Function to read Markdown file content
def read_markdown_file(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        return file.read()
```

```python
# Step 3: Function to convert Markdown text to HTML
def convert_markdown_to_html(markdown_text):
    return markdown.markdown(markdown_text)
```

```python
# Step 4: Function to wrap converted HTML inside a complete HTML template
def wrap_in_html_template(content):
    html_template = f"""
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>Markdown to HTML</title>
        <style>
            body {{
                font-family: Arial, sans-serif;
                line-height: 1.6;
                margin: 20px;
                background-color: #fafafa;
                color: #222;
            }}
            h1, h2, h3 {{
                color: #333;
            }}
            a {{
                color: #1a73e8;
                text-decoration: none;
            }}
            a:hover {{
                text-decoration: underline;
            }}
            code {{
                background-color: #f4f4f4;
                padding: 2px 5px;
                border-radius: 3px;
            }}
        </style>
    </head>
    <body>
        {content}
    </body>
    </html>
    """
    return html_template
```

```python
# Step 5: Function to write the final HTML to a file
def write_html_file(html_content, output_path):
    with open(output_path, "w", encoding="utf-8") as file:
        file.write(html_content)
```

```python
# Step 6: Function to show HTML output directly in notebook or Colab
def display_html_preview(html_content):
    display(HTML(html_content))
```

```python
# Step 7: Main function to run the program
def main():
    print("Welcome to the Markdown to HTML Converter!")
    markdown_file = input("Enter the path to your Markdown file: ")
    output_file = input("Enter the path to save the HTML file: ")

    try:
        # Read and process markdown file
        markdown_text = read_markdown_file(markdown_file)
        html_content = convert_markdown_to_html(markdown_text)
        html_with_template = wrap_in_html_template(html_content)
        write_html_file(html_with_template, output_file)

        print(f" HTML file has been generated and saved to: {output_file}")

        # Display preview
        print("\n📄 Rendering live HTML preview below:")
        display_html_preview(html_with_template)

    except Exception as e:
        print(f" An error occurred: {e}")
```

```python
# Step 8: Run the main program
if __name__ == "__main__":
    main()
```
