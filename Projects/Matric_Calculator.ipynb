# **Matric Calculator**

```python
# Step 1:  Function to get matrix input
import numpy as np
def get_matrix():
    try:
        # Ask user for matrix dimensions
        rows = int(input("Enter the number of rows: "))
        cols = int(input("Enter the number of columns: "))

        # Ask user to input all elements in one line
        print("Enter all matrix elements separated by spaces:")
        numbers = list(map(float, input().split()))

        # Validate number of elements
        if len(numbers) != rows * cols:
            raise ValueError("Number of elements doesn't match matrix size.")

        # Convert list into 2D matrix
        elements = [numbers[i * cols:(i + 1) * cols] for i in range(rows)]
        return np.array(elements)

    except ValueError as e:
        print("Error:", e)
        return None
```

```python
# Step 2: Function to perform matrix operations

def matrix_operations(A, B):
    print("\nMatrix A:\n", A)
    print("\nMatrix B:\n", B)

    # Addition
    try:
        print("\nAddition:\n", A + B)
    except ValueError:
        print("\nAddition: Matrices must have the same dimensions.")

    # Subtraction
    try:
        print("\nSubtraction:\n", A - B)
    except ValueError:
        print("\nSubtraction: Matrices must have the same dimensions.")

    # Element-wise Multiplication
    try:
        print("\nElement-wise Multiplication:\n", A * B)
    except ValueError:
        print("\nElement-wise Multiplication: Matrices must have the same dimensions.")

    # Dot Product
    try:
        print("\nDot Product:\n", np.dot(A, B))
    except ValueError:
        print("\nDot Product: Columns of A must equal rows of B.")

    # Transpose
    print("\nTranspose of A:\n", A.T)
    print("\nTranspose of B:\n", B.T)

    # Determinant of A
    try:
        print("\nDeterminant of A:", np.linalg.det(A))
    except np.linalg.LinAlgError:
        print("\nDeterminant of A: Not applicable (Matrix must be square).")

    # Inverse of A
    try:
        print("\nInverse of A:\n", np.linalg.inv(A))
    except np.linalg.LinAlgError:
        print("\nInverse of A: Not invertible.")
```

```python
# Step 3: Main program to run calculator
def main():
    print("==========================")
    print("MATRIX CALCULATOR")
    print("==========================")

    print("\nInput Matrix A:")
    A = get_matrix()
    if A is None:
        return

    print("\nInput Matrix B:")
    B = get_matrix()
    if B is None:
        return

    matrix_operations(A, B)
```

```python
# Step 5: Run the program
if __name__ == "__main__":
    main()
```
