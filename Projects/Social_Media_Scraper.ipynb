# **Social Media Scraper**

```python
from bs4 import BeautifulSoup
import csv
import os

# Step 1: Load or create HTML, extract posts, and save to CSV
def main():
    print("=== Social Media Scraper ===")

    # Create a sample HTML file if it doesn't exist
    sample_html = """
    <html>
    <body>
        <div class="post">
            <h2 class="username">User 1</h2>
            <p class="content">This is the first post content!</p>
            <span class="timestamp">2025-01-14 10:00:00</span>
        </div>
        <div class="post">
            <h2 class="username">User 2</h2>
            <p class="content">This is another interesting post!</p>
            <span class="timestamp">2025-01-14 10:05:00</span>
        </div>
    </body>
    </html>
    """
    # Save sample HTML if file doesn't exist
    if not os.path.exists("social_media.html"):
        with open("social_media.html", "w", encoding="utf-8") as f:
            f.write(sample_html)

    # Load HTML file
    with open("social_media.html", "r", encoding="utf-8") as f:
        html_content = f.read()

    # Parse HTML
    soup = BeautifulSoup(html_content, "html.parser")

    # Extract posts
    posts = []
    for post in soup.find_all("div", class_="post"):
        username = post.find("h2", class_="username").text.strip()
        content = post.find("p", class_="content").text.strip()
        timestamp = post.find("span", class_="timestamp").text.strip()
        posts.append({"username": username, "content": content, "timestamp": timestamp})

    # Print extracted posts
    print("\nExtracted Posts:")
    for post in posts:
        print(f"- {post['username']} said: \"{post['content']}\" at {post['timestamp']}")

    # Save posts to CSV
    with open("social_media_posts.csv", "w", newline="", encoding="utf-8") as file:
        writer = csv.DictWriter(file, fieldnames=["username", "content", "timestamp"])
        writer.writeheader()
        writer.writerows(posts)

    print("\n Posts saved successfully to 'social_media_posts.csv'")

# Run the scraper
if __name__ == "__main__":
    main()
```
