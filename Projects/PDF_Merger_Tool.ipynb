```python
import os
from PyPDF2 import PdfMerger

# Step 1: Ask user for the order of PDFs to merge
def get_pdf_order():
    pdf_list = input("Enter PDF file names in the desired order (comma-separated): ").split(",")
    return [pdf.strip() for pdf in pdf_list]
```

```python
# Step 2: Validate if all provided PDF files exist
def validate_files(pdf_list):
    for pdf in pdf_list:
        if not os.path.exists(pdf):
            print(f"Error: {pdf} does not exist.")
            return False
    return True
```

```python
# Step 3: Merge PDFs and save as a new file
def merge_pdfs(pdf_list, output_file):
    merger = PdfMerger()  # Initialize the merger object
    for pdf in pdf_list:
        merger.append(pdf)  # Append each PDF in order
    merger.write(output_file)  # Write to output file
    merger.close()  # Close the merger
    print(f"Merged PDF saved as {output_file}")
```

```python
# Step 4: Main program
def main():
    print("Welcome to the PDF Merger Tool!")

    # Get PDF order from user
    pdf_list = get_pdf_order()

    # Validate files before merging
    if validate_files(pdf_list):
        output_file = input("Enter the output file name (e.g., merged.pdf): ")
        merge_pdfs(pdf_list, output_file)

# Step 5: Run the program
if __name__ == "__main__":
    main()
```
