# **Currency Converter**

```python
# Step 1: Import libraries
import json
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
# Step 2: Load exchange rates
def load_exchange_rates(file_path="/content/exchange_rates.json"):
    try:
        with open(file_path, "r") as file:
            return json.load(file)
    except FileNotFoundError:
        print("Exchange rate file not found! Please upload 'exchange_rates.json'.")
        return {}
```

```python
# Step 3: Function to convert currency
def convert_currency(amount, from_currency, to_currency, rates):
    if from_currency not in rates or to_currency not in rates:
        return None
    base_amount = amount / rates[from_currency]
    return base_amount * rates[to_currency]
```

```python
# Step 4: Function to handle conversion
def handle_conversion(b):
    with result_output:
        clear_output()
        try:
            amount = float(amount_entry.value)
            from_currency = from_currency_dropdown.value
            to_currency = to_currency_dropdown.value
            rates = load_exchange_rates()
            result = convert_currency(amount, from_currency, to_currency, rates)
            if result is not None:
                print(f"{amount} {from_currency} = {result:.2f} {to_currency}")
            else:
                print("Conversion Error! Check currency codes or rates file.")
        except ValueError:
            print("Invalid amount! Please enter a number.")
```

```python
# Step 5: Create widgets
amount_entry = widgets.FloatText(description="Amount:", value=1.0)
from_currency_dropdown = widgets.Dropdown(options=["USD", "EUR", "GBP", "INR", "JPY"], description="From:")
to_currency_dropdown = widgets.Dropdown(options=["USD", "EUR", "GBP", "INR", "JPY"], description="To:")
convert_button = widgets.Button(description="Convert", button_style='success')
result_output = widgets.Output()
```

```python
# Step 6: Bind button click
convert_button.on_click(handle_conversion)
```

```python
# Step 7: Display all widgets
display(amount_entry, from_currency_dropdown, to_currency_dropdown, convert_button, result_output)
```
