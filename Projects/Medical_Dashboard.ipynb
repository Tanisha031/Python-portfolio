# **Medical Dashboard**

```python
# Step 1: Import necessary libraries
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
```

```python
# Step 2: Create input widgets for each metric

# BMI Inputs
weight_input = widgets.FloatText(value=70, description="Weight (kg):", style={'description_width': 'initial'})
height_input = widgets.FloatText(value=1.75, description="Height (m):", style={'description_width': 'initial'})

# Blood Pressure Inputs
systolic_input = widgets.IntText(value=120, description="Systolic (mmHg):", style={'description_width': 'initial'})
diastolic_input = widgets.IntText(value=80, description="Diastolic (mmHg):", style={'description_width': 'initial'})

# Heart Rate Input
heart_rate_input = widgets.IntText(value=70, description="Heart Rate (bpm):", style={'description_width': 'initial'})

# Blood Sugar Input
blood_sugar_input = widgets.FloatText(value=90, description="Blood Sugar (mg/dL):", style={'description_width': 'initial'})

# Output area to display results
output = widgets.Output()
```

```python
# Step 3: Define function to analyze metrics and show advice

def analyze_dashboard(b):
    with output:
        clear_output()

        # --- BMI ---
        weight = weight_input.value
        height = height_input.value
        bmi = weight / (height ** 2)
        if bmi < 18.5:
            bmi_status = "Underweight"
            bmi_color = "orange"
            bmi_advice = "Eat nutritious food & gain weight."
        elif 18.5 <= bmi < 24.9:
            bmi_status = "Normal weight"
            bmi_color = "green"
            bmi_advice = "Maintain your healthy lifestyle."
        elif 25 <= bmi < 29.9:
            bmi_status = "Overweight"
            bmi_color = "orange"
            bmi_advice = "Exercise regularly and control diet."
        else:
            bmi_status = "Obesity"
            bmi_color = "red"
            bmi_advice = "Consult doctor & follow weight management plan."

        # --- Blood Pressure ---
        sys = systolic_input.value
        dia = diastolic_input.value
        if sys < 120 and dia < 80:
            bp_category = "Normal"
            bp_color = "green"
            bp_advice = "Keep up the healthy lifestyle."
        elif 120 <= sys < 130 and dia < 80:
            bp_category = "Elevated"
            bp_color = "#e68a00"
            bp_advice = "Exercise and reduce salt intake."
        elif (130 <= sys < 140) or (80 <= dia < 90):
            bp_category = "Hypertension Stage 1"
            bp_color = "orange"
            bp_advice = "Consult doctor and monitor BP."
        elif sys >= 140 or dia >= 90:
            bp_category = "Hypertension Stage 2"
            bp_color = "red"
            bp_advice = "Seek medical advice immediately."
        else:
            bp_category = "Unusual Reading"
            bp_color = "purple"
            bp_advice = "Check readings again."

        # --- Heart Rate ---
        hr = heart_rate_input.value
        if 60 <= hr <= 100:
            hr_status = "Normal"
            hr_color = "green"
            hr_advice = "Maintain your healthy routine."
        elif hr < 60:
            hr_status = "Bradycardia (Low)"
            hr_color = "orange"
            hr_advice = "Consult doctor if symptomatic."
        else:
            hr_status = "Tachycardia (High)"
            hr_color = "red"
            hr_advice = "Check heart rate and consult doctor."

        # --- Blood Sugar ---
        sugar = blood_sugar_input.value
        if 70 <= sugar <= 99:
            sugar_status = "Normal (Fasting)"
            sugar_color = "green"
            sugar_advice = "Keep up your healthy diet."
        elif 100 <= sugar <= 125:
            sugar_status = "Prediabetes"
            sugar_color = "orange"
            sugar_advice = "Control sugar intake & exercise."
        else:
            sugar_status = "Diabetes"
            sugar_color = "red"
            sugar_advice = "Consult doctor & manage sugar levels."

        # Display all results with advice
        display(HTML(f"""
        <h2 style='color:#4CAF50'> Medical Dashboard Results</h2>
        <p style='color:{bmi_color}'>BMI: {bmi:.2f} ({bmi_status}) - {bmi_advice}</p>
        <p style='color:{bp_color}'>Blood Pressure: {sys}/{dia} mmHg ({bp_category}) - {bp_advice}</p>
        <p style='color:{hr_color}'>Heart Rate: {hr} bpm ({hr_status}) - {hr_advice}</p>
        <p style='color:{sugar_color}'>Blood Sugar: {sugar} mg/dL ({sugar_status}) - {sugar_advice}</p>
        """))
```

```python
# Step 4: Function to reset inputs

def reset_dashboard(b):
    weight_input.value = 70
    height_input.value = 1.75
    systolic_input.value = 120
    diastolic_input.value = 80
    heart_rate_input.value = 70
    blood_sugar_input.value = 90
    with output:
        clear_output()
        display(HTML("<p style='color:gray'>Fields reset to default.</p>"))
```

```python
# Step 5: Buttons for analyzing and resetting

analyze_button = widgets.Button(description="Analyze", button_style="success")
analyze_button.on_click(analyze_dashboard)

reset_button = widgets.Button(description="Reset", button_style="danger")
reset_button.on_click(reset_dashboard)
```

```python
# Step 6: Arrange layout and display dashboard

dashboard_ui = widgets.VBox([
    widgets.HTML("<h1 style='color:#4CAF50'> Multi-Parameter Medical Dashboard</h1>"),
    weight_input,
    height_input,
    systolic_input,
    diastolic_input,
    heart_rate_input,
    blood_sugar_input,
    widgets.HBox([analyze_button, reset_button]),
    output
])

display(dashboard_ui)
```
