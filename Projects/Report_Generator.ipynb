# **Student Report Generator**

```python
import csv

# Step 1: Read the student data
with open('/content/Student.csv', 'r') as infile:
    reader = csv.DictReader(infile)
    students = list(reader)
```

```python
# Step 2: Calculate average and status (pass/fail)
for student in students:
    math = int(student['Math'])
    science = int(student['Science'])
    english = int(student['English'])

    avg = (math + science + english) / 3
    student['Average'] = round(avg, 2)

    # Passing criteria: average >= 50
    student['Status'] = 'Pass' if avg >= 50 else 'Fail'
```

```python
# Step 3: Write the report to a new CSV file
with open('student_report.csv', 'w', newline='') as outfile:
    fieldnames = ['Name', 'Math', 'Science', 'English', 'Average', 'Status']
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(students)
```

```python
# Step 4: Display report summary in console
print("Student Report Generated Successfully!\n")
print("Name\tAverage\tStatus")
for s in students:
    print(f"{s['Name']}\t{s['Average']}\t{s['Status']}")
```

```python
# Step 5: Program complete
print("\nReport saved as 'student_report.csv'")
```
