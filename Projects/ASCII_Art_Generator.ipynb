# **ASCII Art Generator**

```python
# Step 1: Import necessary library
from PIL import Image
```

```python
# Step 2: Load and resize the image
def load_image(image_path, new_width=100):
    # Open the image file
    img = Image.open(image_path)

    # Calculate aspect ratio (height / width)
    aspect_ratio = img.height / img.width

    # Calculate new height (adjusted for text aspect ratio)
    new_height = int(new_width * aspect_ratio * 0.55)

    # Resize the image to new dimensions
    img = img.resize((new_width, new_height))

    return img
```

```python
# Step 3: Convert the image to grayscale
def convert_to_grayscale(img):
    # "L" mode converts the image to 8-bit grayscale
    return img.convert("L")
```

```python
# Step 4: Map grayscale pixels to ASCII characters
def map_pixels_to_ascii(img):
    # Define ASCII characters (from dark to light)
    ascii_chars = "@%#*+=-:. "

    # Get pixel brightness values
    pixels = img.getdata()

    # Convert each pixel to a character
    ascii_str = "".join([ascii_chars[pixel // 25] for pixel in pixels])

    return ascii_str
```

```python
# Step 5: Generate full ASCII art text
def generate_ascii_art(image_path, new_width=100):
    # Load and resize the image
    img = load_image(image_path, new_width)

    # Convert to grayscale
    gray_image = convert_to_grayscale(img)

    # Convert pixels to ASCII string
    ascii_str = map_pixels_to_ascii(gray_image)

    # Format ASCII string into multiple rows
    ascii_art = "\n".join([ascii_str[i:i + new_width] for i in range(0, len(ascii_str), new_width)])

    return ascii_art
```

```python
# Step 6: Save ASCII art to a text file
def save_ascii_art(ascii_art, output_path):
    # Open file and write ASCII text
    with open(output_path, "w", encoding="utf-8") as file:
        file.write(ascii_art)
```

```python
# Step 7: Get user input and handle ASCII generation
def main():
    print("Welcome to the ASCII Art Generator!")

    # Get image path from user
    image_path = input("Enter the path to your image: ").strip()

    # Get output file path from user
    output_path = input("Enter the path to save the ASCII art (e.g., output.txt): ").strip()

    # Get desired width (default 100)
    width_input = input("Enter the desired width of the ASCII art (default is 100): ").strip()
    new_width = int(width_input) if width_input.isdigit() else 100

    try:
        # Generate ASCII art
        ascii_art = generate_ascii_art(image_path, new_width)

        # Save ASCII art to output file
        save_ascii_art(ascii_art, output_path)

        # Confirm completion
        print(f"ASCII art generated and saved to {output_path}")

    except Exception as e:
        # Handle errors
        print(f"An error occurred: {e}")
```

```python
# Step 8: Main program loop
if __name__ == "__main__":
    main()
```
