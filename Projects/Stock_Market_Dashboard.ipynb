# **Stock Market Dashboard**

```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
# Step 1: Load Stock Data
def load_data(file_path="/content/stock_data.csv"):
    return pd.read_csv(file_path)

data = load_data()
```

```python
# Step 2: Function to plot stock data
def plot_stock_data(button):
    clear_output(wait=True)  # Clear previous plot
    stock = stock_dropdown.value
    filtered_data = data[data["stock"] == stock]
    plt.figure(figsize=(8, 5))
    plt.plot(filtered_data["date"], filtered_data["price"], marker="o")
    plt.title(f"{stock} Stock Prices")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
    display(stock_dropdown, plot_button)  # Keep widgets visible after plotting
```

```python
# Step 3: Create dropdown widget
stock_dropdown = widgets.Dropdown(
    options=data["stock"].unique().tolist(),
    description="Select Stock:"
)
```

```python
# Step 4: Create plot button
plot_button = widgets.Button(
    description="Plot Stock Data",
    button_style='success',
    tooltip='Click to plot selected stock'
)
```

```python
# Step 5: Attach function to button click
plot_button.on_click(plot_stock_data)
```

```python
# Step 6: Display widgets
display(stock_dropdown, plot_button)
```
