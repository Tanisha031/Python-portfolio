# **Automated Backup Tool**

```python
import os
import shutil
from datetime import datetime

# Step 1: List files in the source directory
def list_files(directory):
    return [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
```

```python
# Step 2: Copy a file to destination
def copy_file(source, destination):
    shutil.copy2(source, destination)
```

```python
# Step 3: Create a timestamped backup folder
def create_backup_directory(base_backup_dir):
    timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    backup_dir = os.path.join(base_backup_dir, f"backup_{timestamp}")
    os.makedirs(backup_dir, exist_ok=True)
    return backup_dir
```

```python
# Step 4: Backup all files from source to backup folder
def backup_files(source_dir, backup_dir):
    files = list_files(source_dir)
    for file in files:
        source_path = os.path.join(source_dir, file)
        destination_path = os.path.join(backup_dir, file)
        copy_file(source_path, destination_path)
        print(f"Backed up: {file}")
    return files
```

```python

# Step 5: Write the log file inside the backup folder automatically
def write_log(backup_dir, files):
    log_file = os.path.join(backup_dir, "backup_log.txt")
    with open(log_file, "w") as log:
        log.write(f"Backup created: {backup_dir}\n")
        for file in files:
            log.write(f"  - {file}\n")
    print(f"\n Backup log saved to {log_file}")
```

```python
# Step 6: Main program
def main():
    print("Welcome to the Automated Backup Tool!")
    source_dir = input("Enter the source directory: ")
    base_backup_dir = input("Enter the base backup directory: ")

    # Validate source directory
    if not os.path.exists(source_dir):
        print("Source directory does not exist!")
        return

    # Create base backup directory if not exists
    os.makedirs(base_backup_dir, exist_ok=True)

    # Create backup folder and copy files
    backup_dir = create_backup_directory(base_backup_dir)
    files = backup_files(source_dir, backup_dir)

    # Write log automatically inside the backup folder
    write_log(backup_dir, files)

    print(f"\nBackup completed successfully! All files saved in {backup_dir}")

# Step 7: Run the program
if __name__ == "__main__":
    main()
```
