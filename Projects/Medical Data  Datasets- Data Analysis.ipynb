# **Medical** **Data** **Analysis** **with** **Python**

**Step** **01**: **Import** **Libraries** and read CSV file.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step** **02**: **Import** **CSV** **file**.

```python
df = pd.read_csv("/content/medical_records.csv")
```

**Step** **03**: **Exploring the Data and Understanding the Data Structure**

```python
# 1 - Display the first few rows of the dataset
print(df.head())
```

```python
# 2 - List the names of all columns in the DataFrame
print(list(df.columns))
```

```python
# 3 - Number of rows and columns
print(df.shape)
```

```python
# 4 - Data types for each column
df.dtypes
```

```python
# 5 - Number of null values in each column
df.isnull().sum()
```

**Step** **04**: **Understanding** **the** **Statistics**

```python
# Calculate and display various statistics about the data
print(df.describe(include = 'all'))
```

**Step** **05**: **Filtering** **Data** **for** **Stroke** **Patients**

```python
# Filter the data to include only patients diagnosed with Stroke:

stroke_data = df[df['Disease_Category'] == 'Stroke']
print("Stroke data summary:")
print(stroke_data.describe(include='all'))
```

**Step** **06**: **Grouping** **and** **Analyzing** **by** **Disease** **Category**

```python
# Sorting patients by their disease type.
disease_stats = df.groupby('Disease_Category').describe(include='all')
print("Statistics by Disease Category:")
print(disease_stats)
```

**Step** **07**: **Exploratory** **Data** **Analysis**

```python
# 1. Visualizing Age Distribution by Disease Category (Box Plot)
sns.boxplot( x = "Disease_Category", y = "Age", data=df, color = 'magenta')
plt.title('Distribution of Age by Disease Category')
plt.xlabel('Disease Category')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Visualizing Age Distribution by Disease Category (Violin Plot)
sns.violinplot( x = "Disease_Category", y = "Age", data=df, color =  'cyan')
plt.title('Distribution of Age by Disease Category (Violin Plot)')
plt.xlabel('Disease Category')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.show()
```
