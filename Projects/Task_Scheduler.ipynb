# **Task Scheduler**

```python
import json
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
# Step 1: Load and save tasks
def load_tasks(file_path="tasks.json"):
    try:
        with open(file_path, "r") as file:
            return json.load(file)
    except FileNotFoundError:
        return []

def save_tasks(tasks, file_path="tasks.json"):
    with open(file_path, "w") as file:
        json.dump(tasks, file, indent=4)

tasks = load_tasks()
```

```python
# Step 2: Display tasks in a widget
task_output = widgets.Output()

def update_task_list():
    with task_output:
        clear_output()
        for idx, task in enumerate(tasks):
            status = "✔" if task["completed"] else "✘"
            print(f"{idx+1}. {task['title']} (Due: {task['due_date']}) [{status}]")

update_task_list()
```

```python
# Step 3: Widgets for input
title_entry = widgets.Text(description="Task Title:")
date_entry = widgets.Text(description="Due Date:")
```

```python
# Step 4: Buttons
add_button = widgets.Button(description="Add Task", button_style='success')
complete_button = widgets.Button(description="Mark Completed", button_style='info')
delete_button = widgets.Button(description="Delete Task", button_style='danger')
task_selector = widgets.IntText(value=1, description="Task #:")
```

```python
# Step 5: Button functionality
def add_task(b):
    title = title_entry.value.strip()
    due_date = date_entry.value.strip()
    if title and due_date:
        tasks.append({"title": title, "due_date": due_date, "completed": False})
        save_tasks(tasks)
        update_task_list()
        title_entry.value = ""
        date_entry.value = ""

def mark_task_completed(b):
    idx = task_selector.value - 1
    if 0 <= idx < len(tasks):
        tasks[idx]["completed"] = True
        save_tasks(tasks)
        update_task_list()

def delete_task(b):
    idx = task_selector.value - 1
    if 0 <= idx < len(tasks):
        tasks.pop(idx)
        save_tasks(tasks)
        update_task_list()

add_button.on_click(add_task)
complete_button.on_click(mark_task_completed)
delete_button.on_click(delete_task)
```

```python
# Step 6: Display everything
display(title_entry, date_entry, add_button, task_selector, complete_button, delete_button, task_output)
```
