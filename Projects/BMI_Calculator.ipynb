# **BMI Calculator**

```python
# Step 1: Welcome message
print("====================================")
print("      Welcome to BMI Calculator     ")
print("====================================\n")

# Step 2: Get user input for weight and height
# Use a loop to make sure user enters valid numbers
while True:
    try:
        weight = float(input("Step 2: Enter your weight in kilograms (kg): "))
        if weight <= 0:
            print("Weight must be a positive number. Please try again.")
            continue
        break
    except ValueError:
        print("Invalid input! Please enter a numeric value for weight.")

while True:
    try:
        height = float(input("Step 3: Enter your height in meters (m): "))
        if height <= 0:
            print("Height must be a positive number. Please try again.")
            continue
        break
    except ValueError:
        print("Invalid input! Please enter a numeric value for height.")

# Step 4: Calculate BMI using the formula
# BMI = weight (kg) / (height (m))^2
bmi = weight / (height ** 2)
bmi = round(bmi, 2)

# Step 5: Determine BMI category based on standard ranges
if bmi < 18.5:
    category = "Underweight"
elif 18.5 <= bmi < 24.9:
    category = "Normal weight"
elif 25 <= bmi < 29.9:
    category = "Overweight"
else:
    category = "Obese"

# Step 6: Display the results in a readable format
print("\n====================================")
print(f"Step 6: Calculated BMI: {bmi}")
print(f"You are classified as: {category}")
print("====================================\n")

# Step 7: Optional advice based on category
if category == "Underweight":
    print("Advice: You may need to gain some weight. Consider consulting a nutritionist.")
elif category == "Normal weight":
    print("Advice: Great! Keep maintaining your healthy weight.")
elif category == "Overweight":
    print("Advice: Consider a balanced diet and regular exercise to lose weight.")
else:
    print("Advice: You may need to reduce your weight for better health. Consult a doctor or nutritionist.")

print("\nThank you for using the BMI Calculator!")
```
