# **Personal Diary App**

```python
# Step 1: Import required libraries
import os
import getpass
from datetime import datetime
from cryptography.fernet import Fernet
```

```python
# Step 2: Encryption setup functions

# Generate and save a key if it doesn't exist
def generate_key():
    if not os.path.exists("secret.key"):
        key = Fernet.generate_key()
        with open("secret.key", "wb") as key_file:
            key_file.write(key)

# Load the saved key
def load_key():
    return open("secret.key", "rb").read()

# Encrypt a text string
def encrypt_text(text):
    key = load_key()
    cipher = Fernet(key)
    return cipher.encrypt(text.encode())

# Decrypt an encrypted string
def decrypt_text(encrypted_text):
    key = load_key()
    cipher = Fernet(key)
    return cipher.decrypt(encrypted_text).decode()
```

```python
# Step 3: Diary entry functions

# Create a new encrypted diary entry
def create_entry():
    title = input("Enter the title of your diary entry: ")
    content = input("Enter your diary content: ")
    date = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")

    # Encrypt content
    encrypted_content = encrypt_text(content)

    # Ensure entries folder exists
    os.makedirs("entries", exist_ok=True)

    # Save entry to file
    file_name = f"{date}_{title}.txt"
    with open(os.path.join("entries", file_name), "wb") as file:
        file.write(encrypted_content)
    print(f"Diary entry '{title}' saved successfully!")

# List all diary entries
def list_entries():
    os.makedirs("entries", exist_ok=True)
    entries = os.listdir("entries")
    if entries:
        print("Your Diary Entries:")
        for index, entry in enumerate(entries, start=1):
            print(f"{index}. {entry}")
    else:
        print("No diary entries found.")

def read_entry():
    # List all entries
    os.makedirs("entries", exist_ok=True)
    entries = os.listdir("entries")

    if not entries:
        print("No diary entries found.")
        return

    print("Your Diary Entries:")
    for index, entry in enumerate(entries, start=1):
        print(f"{index}. {entry}")

    # Ask user for the entry number instead of typing full name
    try:
        choice = int(input("Enter the number of the entry to read: "))
        if 1 <= choice <= len(entries):
            file_name = entries[choice - 1]
            file_path = os.path.join("entries", file_name)

            with open(file_path, "rb") as file:
                encrypted_content = file.read()
            content = decrypt_text(encrypted_content)

            print("\nDiary Entry Content:")
            print(content)
        else:
            print("Invalid number selected.")
    except ValueError:
        print("Please enter a valid number.")
```

```python
# Step 4: Authentication function

def authenticate():
    correct_password = "PassW0rd"  # Set your password here
    password = getpass.getpass("Enter your password: ")
    if password == correct_password:
        print("Access Granted!")
        return True
    else:
        print("Access Denied!")
        return False
```

```python
# Step 5: Main program loop
def main():
    generate_key()  # Ensure encryption key exists
    if authenticate():  # Verify password
        while True:
            print("\nOptions:")
            print("1. Create a New Entry")
            print("2. View All Entries")
            print("3. Read an Entry")
            print("4. Exit")
            choice = input("Enter your choice: ")
            if choice == "1":
                create_entry()
            elif choice == "2":
                list_entries()
            elif choice == "3":
                read_entry()
            elif choice == "4":
                print("Goodbye!")
                break
            else:
                print("Invalid choice. Please try again.")
```

```python
# Step 6: Run the program
if __name__ == "__main__":
    main()
```
