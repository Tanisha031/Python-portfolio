# **Stock Prize Tracker**

```python
# Step 1: Import Libraries
import requests
from bs4 import BeautifulSoup
import time
import yfinance as yf
from datetime import datetime
```

```python
# Step 2: Fetch the HTML content of a page
def fetch_page(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to fetch page: {response.status_code}")
        return None
```

```python
# Step 3: Parse HTML using BeautifulSoup
def parse_html(html):
    soup = BeautifulSoup(html, "html.parser")
    return soup
```

```python
# Step 4: Get the stock price for a specific ticker
def get_stock_price(ticker):
    url = f"https://finance.yahoo.com/quote/{ticker}"
    html = fetch_page(url)
    if not html:
        return None

    soup = parse_html(html)
    price_tag = soup.find("fin-streamer", {"data-symbol": ticker, "data-field": "regularMarketPrice"})
    if price_tag:
        return price_tag.text
    else:
        print("Stock price not found.")
        return None
```

```python
# Step 5: Function to fetch and display stock price
def track_stock_price(ticker, interval=60, updates=10):
    print(f"\nTracking {ticker} for {updates} updates every {interval} seconds...\n")
    for i in range(updates):
        try:
            stock = yf.Ticker(ticker)
            price = stock.info.get("regularMarketPrice")
            if price is not None:
                print(f"[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] {ticker}: ${price}")
            else:
                print(f"[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] Could not fetch price for {ticker}")
        except Exception as e:
            print("Error fetching data:", e)
        time.sleep(interval)
    print("\nTracking completed.")
```

```python
# Step 6: Main Program
def main():
    print("Welcome to the Stock Price Tracker!")
    ticker = input("Enter the stock ticker symbol (e.g., AAPL, TSLA): ").upper()
    interval = int(input("Enter the update interval (in seconds): "))
    updates = int(input("Enter the number of updates to track: "))
    track_stock_price(ticker, interval, updates)
```

```python
# Main Execution
if __name__ == "__main__":
    main()
```
