# **Expense Tracker**

```python
# Step 1: Import the Libraries
import pandas as pd
import matplotlib.pyplot as plt
import csv
from datetime import datetime
```

```python
# Step 2: Log an Expense
def log_expense(date, category, amount, description):
    with open('expenses.csv', 'a', newline="") as file:
        writer = csv.writer(file)
        writer.writerow([date, category, amount, description])
```

```python
# Step 3: Load Expenses
def load_expenses():
    return pd.read_csv("expenses.csv", names=["Date", "Category", "Amount", "Description"])
```

```python
# Step 4:  Summarize Expenses
def summarize_expenses(df):
    summary = df.groupby("Category")["Amount"].sum()
    print("\nExpense Summary:")
    print(summary)
```

```python
# Step 5: Plot Expenses by Category
def plot_expenses_by_category(df):
    summary = df.groupby("Category")["Amount"].sum()
    summary.plot(kind="pie", autopct="%1.1f%%", figsize=(8,8), title="Expenses by Category")
    plt.ylabel("")
    plt.show(block=False)  # Non-blocking
    plt.pause(3)           # Display for 3 seconds
    plt.close()            # Close automatically
```

```python
# Step 6: Plot Monthly Trends
def plot_monthly_trends(df):
    df["Date"] = pd.to_datetime(df["Date"])
    df["Month"] = df["Date"].dt.to_period("M")
    monthly_summary = df.groupby("Month")["Amount"].sum()
    monthly_summary.plot(kind="bar", figsize=(10, 6), title="Monthly Expense Trends")
    plt.xlabel("Month")
    plt.ylabel("Total Expenses")
    plt.xticks(rotation=45)
    plt.show(block=False)  # Non-blocking
    plt.pause(3)           # Display for 3 seconds
    plt.close()            # Close automatically
```

```python
# Step 7: Main Menu
def main():
    print("Welcome to the Expense Tracker!")
    while True:
        print("\nOptions:")
        print("1. Log an Expense")
        print("2. View Expense Summary")
        print("3. Plot Expenses by Category")
        print("4. Plot Monthly Trends")
        print("5. Exit")

        choice = input("Enter your choice: ")
        if choice == "1":
            date = input("Enter date (YYYY-MM-DD): ")
            category = input("Enter category (e.g., Food, Transport, Shopping): ")
            amount = float(input("Enter amount: "))
            description = input("Enter description: ")
            log_expense(date, category, amount, description)
            print("Expense logged successfully!")
        elif choice == "2":
            df = load_expenses()
            summarize_expenses(df)
        elif choice == "3":
            df = load_expenses()
            plot_expenses_by_category(df)
        elif choice == "4":
            df = load_expenses()
            plot_monthly_trends(df)
        elif choice == "5":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")
```

```python
# Step 8: Run the Program

if __name__ == "__main__":
    main()
```
