# **Tic-Tac-Toe Game**

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import random

# Step 1: Initialize board and player
board = [["" for _ in range(3)] for _ in range(3)]
player_symbol = "X"   # You
computer_symbol = "O" # Computer
```

```python
# Step 2: Output widget for status
status = widgets.Label(value="Your Turn (X)")
```

```python
# Step 3: Functions for game logic
def check_winner(board):
    # Rows
    for row in board:
        if row[0] == row[1] == row[2] and row[0] != "":
            return row[0]
    # Columns
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] and board[0][col] != "":
            return board[0][col]
    # Diagonals
    if board[0][0] == board[1][1] == board[2][2] and board[0][0] != "":
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] and board[0][2] != "":
        return board[0][2]
    return None

def is_draw(board):
    for row in board:
        if "" in row:
            return False
    return True

def disable_buttons():
    for row in buttons:
        for b in row:
            b.disabled = True
```

```python
# Step 4: Computer Move (random)
def computer_move():
    empty_cells = [(r, c) for r in range(3) for c in range(3) if board[r][c] == ""]
    if empty_cells:
        r, c = random.choice(empty_cells)
        board[r][c] = computer_symbol
        buttons[r][c].description = computer_symbol
        winner = check_winner(board)
        if winner:
            status.value = f"{winner} Wins!"
            disable_buttons()
            return True
        elif is_draw(board):
            status.value = "It's a Draw!"
            return True
    return False
```

```python
# Step 5: Player move
def on_click(r, c):
    if board[r][c] == "":
        board[r][c] = player_symbol
        buttons[r][c].description = player_symbol

        winner = check_winner(board)
        if winner:
            status.value = f"{winner} Wins!"
            disable_buttons()
        elif is_draw(board):
            status.value = "It's a Draw!"
        else:
            status.value = "Computer's Turn (O)"
            if not computer_move():
                status.value = "Your Turn (X)"
```

```python
# Step 6: Reset Game
def reset_game(_):
    global board
    board = [["" for _ in range(3)] for _ in range(3)]
    for row in buttons:
        for b in row:
            b.description = ""
            b.disabled = False
    status.value = "Your Turn (X)"
```

```python
# Step 7: Create Buttons
buttons = []
for i in range(3):
    row = []
    for j in range(3):
        b = widgets.Button(description="", layout=widgets.Layout(width='60px', height='60px'))
        b.on_click(lambda btn, r=i, c=j: on_click(r, c))
        row.append(b)
    buttons.append(row)
```

```python
# Step 8: Display Board
board_box = widgets.VBox([widgets.HBox(row) for row in buttons])
reset_button = widgets.Button(description="Reset", button_style='info')
reset_button.on_click(reset_game)

display(status)
display(board_box)
display(reset_button)
```
