# **Pomodoro Timer**

```python
# Step 1: Import necessary libraries
import time
import threading
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
# Step 2: Initialize global variables
session_count = 0
timer_running = False
stop_timer = False
```

```python
# Step 3: Create UI elements
start_button = widgets.Button(description="Start Timer", button_style='success')
reset_button = widgets.Button(description="Reset Timer", button_style='warning')
exit_button = widgets.Button(description="Exit", button_style='danger')

status_label = widgets.Label(value="Status: Ready")
timer_label = widgets.Label(value="25:00")
```

```python
# Step 4: Define countdown function
def countdown(seconds):
    global timer_running, stop_timer
    timer_running = True
    while seconds >= 0 and not stop_timer:
        mins, secs = divmod(seconds, 60)
        timer_label.value = f"{mins:02d}:{secs:02d}"
        time.sleep(1)
        seconds -= 1
    timer_running = False
    if not stop_timer:
        next_session()
```

```python
# Step 5: Define start timer logic
def start_timer(_=None):
    global session_count, stop_timer
    if not timer_running:
        stop_timer = False
        session_count += 1

        if session_count % 8 == 0:
            status_label.value = "Status: Long Break (15 min)"
            threading.Thread(target=countdown, args=(15 * 60,), daemon=True).start()
        elif session_count % 2 == 0:
            status_label.value = "Status: Short Break (5 min)"
            threading.Thread(target=countdown, args=(5 * 60,), daemon=True).start()
        else:
            status_label.value = "Status: Work Session (25 min)"
            threading.Thread(target=countdown, args=(25 * 60,), daemon=True).start()
```

```python
# Step 6: Define reset timer logic
def reset_timer(_=None):
    global session_count, stop_timer
    stop_timer = True
    session_count = 0
    timer_label.value = "25:00"
    status_label.value = "Status: Ready"
```

```python
# Step 7: Define exit logic
def exit_program(_=None):
    global stop_timer
    stop_timer = True
    clear_output()
    print("Pomodoro session ended. Goodbye!")
```

```python
# Step 8: Automatically start next session
def next_session():
    time.sleep(2)
    start_timer()
```

```python
# Step 9: Connect button actions
start_button.on_click(start_timer)
reset_button.on_click(reset_timer)
exit_button.on_click(exit_program)
```

```python
# Step 10: Display UI
ui = widgets.VBox([
    widgets.HTML(value="<h2>Pomodoro Timer (Colab Version)</h2>"),
    timer_label,
    status_label,
    widgets.HBox([start_button, reset_button, exit_button])
])
display(ui)
```
