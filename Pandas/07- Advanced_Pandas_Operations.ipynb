**Finding** **Correlation** **between** **Data**

```python
# Example of simple correlation method

import pandas as pd
df = {
    "Array_1": [30, 70, 100],
    "Array_2": [65.1, 49.50, 30.7]
}
data = pd.DataFrame(df)
print(data.corr())
```

```python
# Creating Sample Dataframe
# importing pandas as pd
import pandas as pd
# Making data frame from the csv file
df = pd.read_csv("/content/drive/MyDrive/nba.csv")
# Printing the first 10 rows of the data frame for visualization
df[:10]
```

```python
# Python Pandas DataFrame corr() Method Examples

# 1 - Find Correlation Among the Columns Using pearson Method

# Select only numeric columns
num_df = df.select_dtypes(include=['number'])
# Correlation using Pearson method
Corr = num_df.corr(method='pearson')
print(Corr)
```

```python
# 2 - Find Correlation Among the Columns Using Kendall Method

# Select only numeric columns
num_df1 = df.select_dtypes(include=['number'])
# Correlation using Pearson method
corr1 = num_df1.corr(method='kendall')
print(corr1)
```

```python
# 3 - Find Correlation Among the Columns Using Spearman Method

# Select only numeric columns
num_df2 = df.select_dtypes(include=['number'])
# Correlation using Pearson method
corr2 = num_df2.corr(method='spearman')
print(corr2)
```

**Basic** **of Time Series Manipulation Using Pandas**

```python
# Create DateTime Values with Pandas
import pandas as pd
from datetime import datetime
import numpy as np

# range_date
rd = pd.date_range(start ='1/1/2025', end ='1/08/2025', freq ='Min')
print(rd)
```

```python
# Determine the Data Type of an Element in the DateTime Range
import pandas as pd
from datetime import datetime
import numpy as np

# range_date
rd = pd.date_range(start ='1/1/2019', end ='1/08/2019', freq ='Min')
print(type(rd[110]))
```

```python
# Convert DateTime elements to String format
import pandas as pd
from datetime import datetime
import numpy as np

# range_Date
rd = pd.date_range(start ='1/1/2019', end ='1/08/2019',freq ='Min')

df = pd.DataFrame(rd, columns =['date'])
df['data'] = np.random.randint(0, 100, size =(len(rd)))

# string_data
s = [str(x) for x in rd]
print(s[1:11])
```

```python
# Accessing Specific DateTime Element
import pandas as pd
import numpy as np

range_data = pd.date_range(start='1/1/2019', end='1/08/2019', freq='Min')

df = pd.DataFrame(range_data, columns=['datetime'])
df['data'] = np.random.randint(0, 100, size=(len(range_data)))

df.set_index('datetime', inplace=True)

filtered_df = df.loc['2019-01-05']
print(filtered_df.iloc[1:11])
```
