**Pandas** - **Series** **and** **DataFrame**

```python
# Data Structures in Pandas
''' A Pandas Series is one-dimensional labeled array capable of holding data of any type
    (integer, string, float, python objects, etc.). The axis labels are collectively called index.

   Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns).
# '''
```

```python
#  Example :01 - Creating a Series

import pandas as pd
import numpy as np

# 1. Creating an empty Series
ser = pd.Series()
print("Pandas Series: ", ser)
# 2. Creating Series from a NumPy array
data = np.array(['g', 'e', 'e', 'k', 's'])
ser = pd.Series(data)
print("Pandas Series:\n", ser)

```

```python
# creating DataFrame

import pandas as pd

# 1. Creating an empty DataFrame
df = pd.DataFrame()
print(df)

# 2. Creating a DataFrame from a list
lst = ['pandas', 'numpy', 'matplotlib']
df = pd.DataFrame(lst)
print(df)
```
