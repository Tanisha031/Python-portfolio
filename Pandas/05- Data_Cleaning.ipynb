1- Checking Missing Values in Pandas

```python
# Using isnull()

# Example 1: Finding Missing Values in a DataFrame
import pandas as pd
import numpy as np

d = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(d)

mv = df.isnull()

print(mv)
```

2 - Filling Missing Values in Pandas

```python
# 1. Using fillna()

# Example 1: Fill Missing Values with Zero
import pandas as pd
import numpy as np

d = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(d)
df.fillna(0)
```

```python
# Example 2: Fill with Previous Value (Forward Fill)
import pandas as pd
import numpy as np

d = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(d)
df.fillna(method='pad')
```

```python
df.ffill()     # df.ffill() → Forward fill (same as fillna(method='pad'))
```

```python
# Example 3: Fill with Next Value (Backward Fill)
import pandas as pd
import numpy as np

d = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(d)

df.fillna(method='bfill')
```

```python
df.bfill()           # df.bfill() → Backward fill (same as fillna(method='bfill'))
```

3 - Dropping Missing Values in Pandas

```python
# 1. Dropping Rows with At Least One Null Value

import pandas as pd
import numpy as np

dict = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, np.nan, 45, 56],
        'Third Score': [52, 40, 80, 98],
        'Fourth Score': [np.nan, np.nan, np.nan, 65]}
df = pd.DataFrame(dict)

df.dropna()
```

```python
# 2. Dropping Rows with All Null Values

dict = {'First Score': [100, np.nan, np.nan, 95],
        'Second Score': [30, np.nan, 45, 56],
        'Third Score': [52, np.nan, 80, 98],
        'Fourth Score': [np.nan, np.nan, np.nan, 65]}
df = pd.DataFrame(dict)

df.dropna(how='all')
```

```python
# 3. Dropping Columns with At Least One Null Value
dict = {'First Score': [100, np.nan, np.nan, 95],
        'Second Score': [30, np.nan, 45, 56],
        'Third Score': [52, np.nan, 80, 98],
        'Fourth Score': [60, 67, 68, 65]}
df = pd.DataFrame(dict)
df.dropna(axis=1)
```

```python
# 4. Dropping Rows with Missing Values in CSV Files

import pandas as pd
d = pd.read_csv("/content/drive/MyDrive/employees.csv")

nd = d.dropna(axis=0, how='any')

print("Old data frame length:", len(d))
print("New data frame length:", len(nd))
print("Rows with at least one missing value:", (len(d) - len(nd)))
```

**Removing** **Duplicates**

```python
# basic example to see how drop_duplicates() works.

import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Alice", "David"],
    "Age": [25, 30, 25, 40],
    "City": ["NY", "LA", "NY", "Chicago"]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

df_cleaned = df.drop_duplicates()
print("\nModified DataFrame (no duplicates)")
print(df_cleaned)
```

```python
# 1. Dropping Duplicates Based on Specific Columns

import pandas as pd
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'SF', 'Chicago']
})
df_cleaned = df.drop_duplicates(subset=["Name"])
print(df_cleaned)
```

```python
# 2. Keeping the Last Occurrence of Duplicates

import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'NY', 'Chicago']
})
df_cleaned= df.drop_duplicates(keep='last')
print(df_cleaned)
```

```python
# 3. Dropping All Duplicates
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'NY', 'Chicago']
})
df_cleaned = df.drop_duplicates(keep=False)
print(df_cleaned)
```

```python
# 4. Modifying the Original DataFrame Directly

import pandas as pd
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'NY', 'Chicago']
})
df.drop_duplicates(inplace=True)
print(df)
```

```python
# 5. Dropping Duplicates Based on Partially Identical Columns

import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Alice", "David", "Bob"],
    "Age": [25, 30, 25, 40, 30],
    "City": ["NY", "LA", "NY", "Chicago", "LA"]
}

df = pd.DataFrame(data)
df_cleaned = df.drop_duplicates(subset=["Name", "City"])
print(df_cleaned)
```

**Pandas** **Change** **Datatype**

```python
# 1. Using astype() method

import pandas as pd

data = {'Name': ['John', 'Alice', 'Bob', 'Eve', 'Charlie'],
        'Age': [25, 30, 22, 35, 28],
        'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'Salary': [50000, 55000, 40000, 70000, 48000]}

df = pd.DataFrame(data)
# Convert 'Age' column to float type
df['Age'] = df['Age'].astype(float)
print(df.dtypes)
```

**Drop** **Empty** **Columns** **in** **DataFrame**

```python
# Create a Sample DataFrame:

import numpy as np
import pandas as pd

df = pd.DataFrame({'FirstName': ['Vipul', 'Ashish', 'Milan'],
                      "Gender": ["", "", ""],
                         "Age": [0, 0, 0]})
df['Department'] = np.nan

print(df)
```

```python
# Example 1: Remove All Null Value Columns

import numpy as np
import pandas as pd

df = pd.DataFrame({'FirstName': ['Vipul', 'Ashish', 'Milan'],
                            "Gender": ["", "", ""],
                            "Age": [0, 0, 0]})

df['Department'] = np.nan
display(df)

df.dropna(how='all', axis=1, inplace=True)
display(df)             #This will remove Department, since it is fully NaN.
```

```python
# Example 2: Replace Empty Strings with Null and then Drop Null Columns
import numpy as np
import pandas as pd

df = pd.DataFrame({'FirstName': ['Vipul', 'Ashish', 'Milan'],
                            "Gender": ["", "", ""],
                            "Age": [0, 0, 0]})
df['Department'] = np.nan
display(df)
nan_value = float("NaN")
df.replace("",nan_value, inplace=True)    #Now Gender column will also be removed (since it was all empty).

df.dropna(how='all', axis=1, inplace=True)
display(df)
```

```python
# Example 3: Replace Zeros with Null and Drop Null Columns

import numpy as np
import pandas as pd

df = pd.DataFrame({'FirstName': ['Vipul', 'Ashish', 'Milan'],
                            "Gender": ["", "", ""],
                            "Age": [0, 0, 0]})

df['Department'] = np.nan
display(df)

df.replace(0, np.nan, inplace=True)

df.dropna(how='all', axis=1, inplace=True)

display(df)
```

```python
# Example 4: Replace Both Zeros and Empty Strings with Null and Drop Null Columns

import numpy as np
import pandas as pd

df = pd.DataFrame({'FirstName': ['Vipul', 'Ashish', 'Milan'],
                            "Gender": ["", "", ""],
                            "Age": [0, 0, 0]})

df['Department'] = np.nan
display(df)

df.replace(["", 0], np.nan, inplace=True)

df.dropna(how='all', axis=1, inplace=True)

display(df)
```

**Detect Mixed Data Types and Fix it**

```python
# Example of mixed datatype:
import pandas as pd
data_frame = pd.DataFrame( [['tom', 10], ['nick', '15'], ['juli', 14.8]], columns=['Name', 'Age'])  # So it becomes a mixed data type column.
```

```python
# to detect mixed datatype: You cannot rely only on info() (because it won’t show mixed properly).
# 'infer_dtype()' checks the type of data in each column.This helps us identify if the column contains 'mixed','string','integer', etc.

print("\nDetecting Data Types:")
for column in data_frame.columns:
    print(column, ":", pd.api.types.infer_dtype(data_frame[column]))
```

```python
# Fixing  Mixed Types: There are 2 main ways:

# 1 - Using astype() function:

# Transforming mixed data types to single data type
data_frame[column] = data_frame[column].astype(int)
# Traverse data frame to detect data types after fix
for column in data_frame.columns:
    print(column,':',pd.api.types.infer_dtype(data_frame[column]))
```

```python
# 2 - Using to_numeric() function:

# Transforming mixed data types to single data type
data_frame[column] = data_frame[column].apply(lambda x: pd.to_numeric(x, errors = 'ignore'))
# Traverse data frame to detect data types after fix
for column in data_frame.columns:
  print(pd.api.types.infer_dtype(data_frame[column]))
```
