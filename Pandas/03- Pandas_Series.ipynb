# **Pandas** **Series**

**Creating** **a** **Pandas** **Series**

```python
# Creating an Empty Pandas Series

import pandas as pd
s = pd.Series()
print(s)
```

```python
# Creating a Series from a NumPy Array

import numpy as np
import pandas as pd

data = np.array(['A', 'n', 'a', 'l', 'y', 's', 't'])
s = pd.Series(data)
print(s)
```

```python
# Creating a Series from a List

data = ['A', 'n', 'a', 'l', 'y', 's', 't']

s = pd.Series(data)
print(s)
```

```python
# Creating a Series from a Dictionary
dict1 = {'a':1, 'b':2, 'c':3}
s = pd.Series(dict1)
print(s)
```

```python
# Creating a Series Using NumPy Functions

# Use NumPy function numpy.linspace() for creating evenly spaced numbers over a specified range.
import numpy as np
import pandas as pd

ser = pd.Series(np.linspace(1, 10, 5))
print(ser)
```

```python
# Use Numpy Function numpy.random.randn() for generating random numbers from a normal distribution.

import numpy as np
import pandas as pd

ser = pd.Series(np.random.randn(10))
print(ser)
```

```python
# Creating a Series Using range()

import pandas as pd
ser = pd.Series(range(5, 15))
print(ser)
```

```python
# Creating a Series Using List Comprehension

import pandas as pd
ser = pd.Series(range(1,20, 3), index = [x for x in 'abcdefg'])
print(ser)
```

**Access** **Elements** **of** **Pandas** **Series**

```python
# Accessing Element from Series with Position


# Accessing the First Element of Series
# import pandas and numpy
import pandas as pd
import numpy as np

# creating simple array
data = np.array(['A', 'n', 'a', 'l', 'y', 's', 't'])
ser = pd.Series(data)
# retrieve the first element
print(ser[0])
```

```python

# import pandas and numpy
import pandas as pd
import numpy as np
# creating simple array
data = np.array(['a', 'n', 'a', 'l', 'y', 's', 't'])

ser = pd.Series(data)
# retrieve the first element
print(ser[:5])
```

```python
# Accessing Last 10 Elements of Series
# import pandas and numpy
import pandas as pd
import numpy as np

# creating simple array
data = np.array(['d', 'a', 't', 'a', 'a', 'n',
                 'a', 'l', 'y', 's', 't', '3', '1'])
ser = pd.Series(data)

# retrieve the first element
print(ser[-10:])
```

**Binary** **operations** **on** **Pandas** **DataFrame** **and** **Series**

```python
#                                           1. Arithmetic Operations on Series

# Adding Two Series
import pandas as pd
s1 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
s2 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])

# Adding the two Series
result = s1 + s2
print(result)
```

```python
#                                           2.Comparison Operations on Series

# Example: Checking Equality

import pandas as pd
s1 = pd.Series([10, 20, 30])
s2 = pd.Series([10, 25, 30])

# Comparing the two Series
result = s1 == s2
print(result)
```

```python
#                                                       3.  Logical Operations on Series

# Logical AND on Series
import pandas as pd
s1 = pd.Series([True, False, True])
s2 = pd.Series([False, False, True])

# Applying logical AND
result = s1 & s2
print(result)
```

```python
# Handling Missing Data in Binary Operations

# Arithmetic with NaN

import pandas as pd
df1 = pd.DataFrame({'A': [1, 2, None], 'B': [4, None, 6]})
df2 = pd.DataFrame({'A': [1, None, 3], 'B': [None, 5, 6]})

# Adding the DataFrames
result = df1 + df2
print(result)
```

**Pandas** **Series** **Index**() **Methods**

```python

import pandas as pd

data = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

# Accessing the index
print("Original Index:", data.index)

# Modifying the index
data.index = ['w', 'x', 'y', 'z']
print("Modified Series:\n", data)
```

```python
# Examples of Pandas Series Index() Attribute

# Example 1. Assigning Duplicate Index Labels
import pandas as pd
series = pd.Series(['New York', 'Chicago', 'Toronto', 'Lisbon'])

# Creating the row axis labels
series.index = ['City 1', 'City 1', 'City 3', 'City 3']
print(series)
```

```python
# Example 2. Retrieving Index Labels
import pandas as pd
Date = ['1/1/2018', '2/1/2018', '3/1/2018', '4/1/2018']
idx_name = ['Day 1', 'Day 2', 'Day 3', 'Day 4']

sr = pd.Series(data = Date,index = idx_name)
print(sr.index)
```

```python
# Example 3. Resetting Index to Default
import pandas as pd

Date = ['1/1/2018', '2/1/2018', '3/1/2018', '4/1/2018']
idx_name = ['Day 1', 'Day 2', 'Day 3', 'Day 4']

sr = pd.Series(data = Date,        # Series Data
              index = idx_name   # Index
              )

# Resetting index to default
sr.reset_index(drop=True, inplace=True)
print(sr)
```
