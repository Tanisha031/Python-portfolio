**Data Normalization with Pandas**

```python
# create a sample dataset using Pandas and visualize it.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame([
    [180000, 110, 18.9, 1400],
    [360000, 905, 23.4, 1800],
    [230000, 230, 14.0, 1300],
    [60000, 450, 13.5, 1500]
], columns=['Col A', 'Col B', 'Col C', 'Col D'])

print(df)

df.plot(kind='bar')
plt.show()
```

```python
# Normalization Techniques in Pandas

# 1. Maximum Absolute Scaling : This makes all values fall between -1 and 1.
x = df.copy()

for column in x.columns:
           x[column] = x[column] / x[column].abs().max()

print(x)

x.plot(kind='bar')
plt.show()
```

```python
# 2. The min-max feature scaling : This makes all values between 0 and 1.
y = df.copy()

for column in y.columns:
    y[column] = (y[column] - y[column].min()) / (y[column].max() - y[column].min())

print(y)
y.plot(kind='bar')
plt.show()
```

```python
# 3. The z-score method : This makes data have; Mean = 0 , Standard deviation = 1

z = df.copy()

for column in z.columns:
    z[column] = (z[column] - z[column].mean()) / z[column].std()

print(z)
z.plot(kind='bar')
plt.show()
```

```python
# Grouping in Pandas : split-apply-combine
import pandas as pd

data = {
    'Item': ['Cake', 'Cake', 'Bread', 'Pastry', 'Cake'],
    'Flavor': ['Chocolate', 'Vanilla', 'Whole Wheat', 'Strawberry', 'Chocolate'],
    'Price': [250, 220, 80, 120, 250]
    }

df = pd.DataFrame(data)
print(df)
```

```python
# 1. Grouping Data by One Column Using groupby()
grouped = df.groupby('Item')
print(grouped)
```

```python
#  find the total price of each item sold:
print(df.groupby('Item')['Price'].sum())
```

```python
# group by Item and Flavor to see how each flavored item sold
print(df.groupby(['Item', 'Flavor'])['Price'].sum())
```

**Pandas DataFrame.groupby() Methods**

```python
# import the library and dataset
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/nba.csv')
df.head()
```

```python
# Example 1: Grouping by a Single Column

team = df.groupby('Team')
print(team.first()) # Let's print the first entries in all the groups formed.
```

```python
# Example 2: Grouping by Multiple Columns

grouping = df.groupby(['Team', 'Position'])
print(grouping.first())
```

```python
# Example 3 : Applying Aggregation with GroupBy
aggregated_data = df.groupby(['Team', 'Position']).agg(total_salary=('Salary', 'sum'), avg_salary=('Salary', 'mean'), player_count=('Name', 'count'))
print(aggregated_data)
```

```python
# Example 4: How to Apply Transformation Methods?

# Rank players within each team by Salary
df['Rank within Team'] = df.groupby('Team')['Salary'].transform(lambda x: x.rank(ascending=False))
print(df)
```

```python
# Example 5 : Filtering Groups Using Filtration Methods

# Filter groups where the average Salary is >= 5 million
filtered_df = df.groupby('Team').filter(lambda x: x['Salary'].mean() >= 1000000)
print(filtered_df)
```

**Different Types of Joins in Pandas**

```python
# we will first make two DataFrames, namely a and b.

# importing pandas
import pandas as pd

# Creating dataframe a
a = pd.DataFrame()
# Creating Dictionary
d = {'id': [1, 2, 10, 12],
     'val1': ['a', 'b', 'c', 'd']}
a = pd.DataFrame(d)
# printing the dataframe a
print('DataFrame a:\n')
print(a)


# Creating dataframe b
b = pd.DataFrame()
# Creating dictionary
d = {'id': [1, 2, 9, 8],
     'val1': ['p', 'q', 'r', 's']}
b = pd.DataFrame(d)

# printing the dataframe b
print('\nDataFrame b:\n')
print(b)
```

```python
# Types of Joins in Pandas

# 1 - Inner Join

inner_join = pd.merge(a, b, on='id', how='inner')
print('Inner Join:\n')
print(inner_join)
```

```python
# 2 - Left Join
left_join = pd.merge(a, b, on='id', how='left')
left = pd.merge(a, b, on='id', how='left')
print('left join:\n')
print(left_join)
```

```python
# 3 - Right Join
right_join = pd.merge(a, b, on='id', how='right')
print('Right Join:\n')
print(right_join)
```

```python
# 4 - Outer Join
outer_join = pd.merge(a, b, on='id', how='outer')
print('Outer Join:\n')
print(outer_join)
```

```python
# 5 - Index Join
index_join = pd.merge(a,b, left_index=True, right_index=True)
print('Index Join:\n')
print(index_join)
```
