Basic Exception Handling (Easy Start)

```python
# Example: Division by zero

n = 10

try:
    result = n / 0   # This will cause an error
except ZeroDivisionError:
    print("Can't divide by zero!")
```

Syntax Error vs Runtime Exception

```python
# Syntax Error (will stop code completely)
# print("Hello world"   # Missing bracket

# Runtime Exception (can be handled)
try:
    n = 10
    result = n / 0
except ZeroDivisionError:
    print("Runtime error handled!")
```

Full Structure (try, except, else, finally)

```python
try:
    n = 5
    result = 100 / n   # Try changing n = 0

except ZeroDivisionError:
    print("Cannot divide by zero!")

else:
    # Runs only if NO error occurs
    print("Result is:", result)

finally:
    # Always runs (important for cleanup)
    print("Execution finished.")
```

Catching Specific Exceptions

```python
try:
    x = int("abc")   # Invalid conversion

except ValueError:
    print("Invalid number!")

except ZeroDivisionError:
    print("Division error!")
```

Multiple Exceptions in One Block

```python
data = ["10", "twenty", 30]

try:
    total = int(data[0]) + int(data[1])  # 'twenty' causes error

except (ValueError, TypeError) as e:
    print("Error occurred:", e)

except IndexError:
    print("Index out of range")
```

Catch-All Exception (Use Carefully )

```python
try:
    result = "100" / 20   # Invalid operation

except:
    print("Something went wrong!")
```

Using else Block

```python
try:
    num = int("50")
    result = 100 / num

except ValueError:
    print("Invalid input")

except ZeroDivisionError:
    print("Cannot divide by zero")

else:
    print("Success! Result =", result)
```

Using finally Block

```python
try:
    file = open("test.txt", "r")   # File may not exist
    content = file.read()

except FileNotFoundError:
    print("File not found!")

finally:
    print("This always runs (cleanup step)")
```

Raising an Exception

```python
def check_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative!")
    return age

try:
    check_age(-5)

except ValueError as e:
    print("Error:", e)
```

Creating Custom Exception

```python
# Custom exception class
class InsufficientFundsError(Exception):
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError("Not enough balance!")
    return balance - amount

try:
    withdraw(100, 150)

except InsufficientFundsError as e:
    print("Transaction failed:", e)
```

Exception Chaining

```python
import json

try:
    try:
        data = '{"name": "John"'   # Invalid JSON
        json.loads(data)

    except json.JSONDecodeError as e:
        raise ValueError("Invalid JSON format") from e

except ValueError as err:
    print("Error:", err)
    print("Original cause:", err.__cause__)
```

Best Practice Example

```python
try:
    num = int(input("Enter a number: "))
    result = 100 / num

except ValueError:
    print("Please enter a valid number")

except ZeroDivisionError:
    print("Cannot divide by zero")

else:
    print("Result:", result)

finally:
    print("Program ended")
```
