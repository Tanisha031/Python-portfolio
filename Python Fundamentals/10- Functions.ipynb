**BASIC FUNCTION**

```python
# Basic function
def greet():
    print("Hello, welcome to Python!")

greet()
```

```python
# Function with parameter
def greet_user(name):
    print("Hello", name)

greet_user("Sarah")
```

```python
# Function with return value
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```

```python
# Function with default parameter
def greet(name="User"):
    print("Hello", name)

greet()
greet("Ali")
```

**FUNCTION WITH MULTIPLE PARAMETERS**

```python
def greet(name, message="Hello", *args, **kwargs):
    print(f"{message}, {name}!")

    if args:
        print("Args:", args)

    if kwargs:
        print("Kwargs:", kwargs)

greet("Sarah", "Hi", 1, 2, 3, city="NY")
```

**MULTIPLE RETURN VALUES**

```python
def operations(a, b):
    return a + b, a - b, a * b, a / b

s, d, p, q = operations(10, 5)
print(s, d, p, q)
```

**SCOPE (LOCAL vs GLOBAL)**

```python
global_var = "I'm global"

def my_function():
    local_var = "I'm local"
    print(global_var)
    print(local_var)

my_function()
print(global_var)
```

**LAMBDA** **FUNCTIONS**

```python
double = lambda x: x * 2
print(double(5))
```

```python
numbers = [1, 5, 3, 9, 2, 6]

doubled = list(map(lambda x: x * 2, numbers))
evens = list(filter(lambda x: x % 2 == 0, numbers))

print(doubled)
print(evens)
```

**RECURSION**

```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

**HIGHER ORDER FUNCTIONS**

```python
def apply_operation(x, y, operation):
    return operation(x, y)

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

print(apply_operation(5, 3, add))
print(apply_operation(5, 3, multiply))
```

**CLOSURE**

```python
def outer(x):
    def inner(y):
        return x + y
    return inner

add_five = outer(5)
print(add_five(3))
print(add_five(7))
```

**ERROR HANDLING IN FUNCTIONS**

```python
def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"
    except TypeError:
        return "Invalid input"

print(divide(10, 2))
print(divide(10, 0))
```

**DECORATORS**

```python
def decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

@decorator
def say_hello():
    print("Hello!")

say_hello()
```

**GENERATORS**

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for i in countdown(5):
    print(i)
```
