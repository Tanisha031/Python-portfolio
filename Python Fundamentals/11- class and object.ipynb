```python
# it will only take one digit value

class mycomplexnumber:
  #constructor method
  def _init_(self, real = 0, imag = 0):
    print("mycomplexnumber Constructor excuting...")
    self.real_part = real
    self.imag_part = imag

  def displayComplex(self):
    print('{0} + {1}j' .format(self.real_part,self.imag_part, 'i'))
```

```python
complxx1 = mycomplexnumber(5,6)
complxx1.displayComplex()
```

```python
#  this program can take number of digits.

class mycomplexnumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def displayComplex(self):
        print(f"{self.real} + {self.imag}i")
```

```python
cmplx1 = mycomplexnumber(31, 15)
cmplx1.displayComplex()
```

```python
complx2 = mycomplexnumber(3, 5)
complx2.displayComplex()
```

```python
complx3 = mycomplexnumber(101, 225)
complx3.displayComplex()
```
