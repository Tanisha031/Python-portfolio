**PYTHON** - **REMOVE** **DICTIONARY** **ITEM**

Using pop() method to remove specified key item

```python
car = {
    'brand' : 'ford',
    'model' : 'mustang',
    'year' : 1964}

car.pop('model')
print(car)
```

popitem() this method will remove thelast inserted item from the list.
(3.7, a random item is removed insteady

```python
car = {
    'brand' : 'ford',
    'model' : 'mustang',
    'year' : 1964}

car.popitem()
print(car)
```

del keyword can delete the specified item and can delete whole dictionary too.

```python
car = {
    'brand' : 'ford',
    'model' : 'mustang',
    'year' : 1964}

del car['model']
print(car)
```

```python
car = {
    'brand' : 'ford',
    'model' : 'mustang',
    'year' : 1964}

del car
print(car)                    #car is not defined , bcoz car dictionary is deleted.
```

clear() method empties the dictionaries

```python
car = {
    'brand' : 'ford',
    'model' : 'mustang',
    'year' : 1964}

car.clear()
print(car)
```

**PYTHON** - **LOOP** **DICTIONARIES**

Loop through a dictionary

```python
mobile = {
    'name' : 'apple',
    'color' : 'red',
    'year' : 2020}

for x in mobile:
  print(x)
```

put all values in dictionary. print one by one

```python
mobile = {
    'name' : 'apple',
    'color' : 'red',
    'year' : 2020}

for x in mobile:
  print(mobile[x])
```

you can also use values to print the same output

```python
mobile = {
    'name' : 'apple',
    'color' : 'red',
    'year' : 2020}

for x in mobile.values():
  print(x)
```

clear()
loop through both to display both key and values in dictionary using item()

```python
mobile = {
    'name' : 'apple',
    'color' : 'red',
    'year' : 2020}

for x , y in mobile.items():
  print(x, y)
```

**PYTHON** - **NESTED** **DICTIONARY**

Create a dictionary that contain dictionaries

```python
myfamily = {
    'child1' : {
        'name' : 'Emil',
        'year' : 2004
    },
    'child2' : {
        'name' : 'Tobias',
        'year' : 2007
    },
    'child3' : {
        'name' : 'linus',
        'year' : 2011
    }
}

print(myfamily)
```

create three dictionaries, the create one dictionary that will contain the other  three dictionaries

```python
child1 = {
    'name' : 'Emil',
    'year' : 2004
}

child2 = {
    'name' : 'Tobias',
    'year' : 2007
}

child3 = {
    'name' : 'linus',
    'year' : 2011
}

myfamily = {
    'child1' : child1,
    'child2' : child2,
    'child3' : child3
}

print(myfamily)
```

access items in nested dictionaries

```python
print(myfamily['child2']['name'])
```

lopp through a key and values in nested dictionaries

```python
for x , obj in myfamily.items():
  print(x)
  for y in obj:
    print(y,':', obj[y])
```
