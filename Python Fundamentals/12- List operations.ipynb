**PYTHON** - **LIST** **COMPREHENSION**

without list comprehension using for loop.

```python
fruits = ['apple', 'banana', 'cherry', 'mango', 'pineapple', 'kiwi', 'grapes']
newlist = []

for x in fruits:
  if 'a' in x:
    newlist.append(x)

print(newlist)
```

with list comprehension

```python
fruits = ['apple', 'banana', 'cherry', 'mango', 'kiwi', 'grapes']
newlist = [x for x in fruits if 'a' in x]
print(newlist)
```

only accept those item who evaulate the condition 'True'

```python
animals = ['cow', 'cat', 'dog', 'rabbit', 'goat']
newlist = [x for x in animals if x!= 'cat']
print(newlist)
```

without if statement

```python
newlist = [x for x in animals]
print(newlist)
```

use 'range' function to create iterable

```python
newlist = [x for x in range(10)]
print(newlist)
```

create iterable but withh condition

```python
newlist = [x for x in range(10) if x < 5]
print(newlist)
```

set the values in uppercase in newlist

```python
newlist =[x.upper() for x in animals]
print(newlist)
```

set all the values in list 'hello'

```python
newlist = ['hello' for x in animals]
print(newlist)
```

return 'mango' instead of 'banana'

```python
fruits_list = ['mango', 'orange', 'banana', 'kiwi']
newlist = [x if x != 'banana' else 'mango' for x in fruits_list]
print(newlist)
```

**PYTHON** - **SORT** **LISTS**

sort lists alphanumerically

```python
fruits = ['banana', 'carrot', 'kiwi', 'mango', 'banana', 'cherry', 'apple']
fruits.sort()
print(fruits)
```

sort the list numericaly

```python
numbers= [31, 1, 5, 32, 21, 6, 25, 29]
numbers.sort()
print(numbers)
```

sort th list descending

```python
numbers.sort(reverse = True)
print(numbers)
fruits.sort(reverse = True)
print(fruits)
```

customize the sort function. using keyword (key = function)

```python
def myfunc(n):
  return abs(n- 50)

thislist = [100, 50, 65, 82, 23]
thislist.sort(key = myfunc)
print(thislist)
```

case sorting results give an unexpected results (prints all capitals letter being sorted before lower case letter)

```python
thislist = ['apple', 'Kiwi', 'orange', 'Banana', 'cherry']
thislist.sort()
print(thislist)
```

use built-in function: if you want case-insensitive use str.lower key function

```python
thislist = ['apple', 'Kiwi', 'Banana', 'orange', 'cherry', 'Grapes']
thislist.sort(key = str.lower)
print(thislist)
```

Reverse the order of list items

```python
thislist.reverse()
print(thislist)
```
