**PYTHON** - **UNPACK** **TUPLE**

unpacking the tuples: extract value back into a variables

```python
fruits = ('apple', 'banana', 'cherry', 'mango')
(green, yellow, red, purple) = fruits

print(green)
print(yellow)
print(red)
print(purple)
```

using *asterik, if number of variable are less than number of values. 'as a list form'

```python
fruits = ('apple', 'banana', 'cherry', 'mango')
(green, yellow, *red) = fruits

print(green)
print(yellow)
print(red)
```

using *asterik, and assign to the second varaible instead of last one item.

```python
fruits = ('apple', 'banana', 'cherry', 'strawberry', 'raspberry')
(green, *tropic, red) = fruits

print(green)
print(tropic)
print(red)
```

PYTHON - LOOP TUPLE

Loop through a tuple

```python
fruits = ('apple', 'banana', 'cherry', 'strawberry', 'raspberry')
for x in fruits:
  print(x)
```

loop through a index number

```python
fruits = ('apple', 'banana', 'cherry', 'strawberry', 'raspberry')
for i in range(len(fruits)):
  print(fruits[i])
```

using a while loop

```python
fruits = ('apple', 'banana', 'cherry', 'mango', 'strawberry')
i = 0
while i < len(fruits):
  print(fruits[i])

  i = i + 1
```
