Open a File (Basic)

```python
# Open file in read mode
file = open("geek.txt", "r")

# Print file object
print(file)

# Always close file
file.close()
```

Check File Properties

```python
file = open("geek.txt", "r")

print("File name:", file.name)
print("Mode:", file.mode)
print("Is closed?", file.closed)

file.close()

print("Is closed after closing?", file.closed)
```

Read Entire File

```python
file = open("geek.txt", "r")

content = file.read()   # Read all content
print(content)

file.close()
```

Read File Line by Line

```python
with open("geek.txt", "r") as file:
    for line in file:
        print(line.strip())   # remove extra newline
```

Read All Lines as List

```python
with open("geek.txt", "r") as file:
    lines = file.readlines()

print(lines)   # list of lines
```

Write to File (Overwrite)

```python
# This will overwrite existing content
with open("geek.txt", "w") as file:
    file.write("Hello Python\n")
    file.write("File handling is easy")

print("File written successfully")
```

Append to File

```python
# This will ADD content at end
with open("geek.txt", "a") as file:
    file.write("\nThis is appended line")
```

Using with

```python
# No need to manually close file
with open("geek.txt", "r") as file:
    content = file.read()
    print(content)

# File automatically closed here
```

Exception Handling with Files

```python
try:
    file = open("geek.txt", "r")
    content = file.read()
    print(content)

except FileNotFoundError:
    print("File not found!")

finally:
    file.close()   # always runs
```

Better Exception Handling (with + try)

```python
try:
    with open("geek.txt", "r") as file:
        content = file.read()
        print(content)

except FileNotFoundError:
    print("File does not exist!")

except PermissionError:
    print("No permission to access file")
```

Working with CSV Files

```python
import csv

# Reading CSV
with open("data.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)   # each row is list


# Writing CSV
with open("output.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Age"])
    writer.writerow(["Ali", 25])
    writer.writerow(["Sara", 22])
```

Working with JSON Files

```python
import json

# Writing JSON
data = {
    "name": "Ali",
    "age": 25
}

with open("data.json", "w") as file:
    json.dump(data, file, indent=4)


# Reading JSON
with open("data.json", "r") as file:
    loaded = json.load(file)
    print(loaded)
```

```python
# Read binary file
with open("image.jpg", "rb") as file:
    data = file.read()

# Write binary file
with open("copy.jpg", "wb") as file:
    file.write(data)
```

Real-World Safe File Handling

```python
def read_file(filename):
    try:
        with open(filename, "r") as file:
            return file.read()

    except FileNotFoundError:
        print("File not found!")

    except Exception as e:
        print("Error:", e)

    return None


content = read_file("geek.txt")
print(content)
```
