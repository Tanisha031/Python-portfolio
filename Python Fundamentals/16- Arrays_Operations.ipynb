NUMPY OPERATIONS

```python
import numpy as np

new_york = np.array([45, 48, 52, 58, 65])
los_angeles = np.array([68, 70, 72, 74, 76])

print(new_york + los_angeles)
print(new_york * 2)

print(np.mean(new_york))
print(np.max(los_angeles))
print(np.min(new_york))
print(np.std(new_york))
```

**RESHAPING** **AND** **SLICING**

```python
import numpy as np

quarterly_sales = np.array([55000, 68000, 72000, 98000, 62000, 75000, 83000, 91000])

sales_by_year = quarterly_sales.reshape(2, 4)
print(sales_by_year)

print(sales_by_year[0])
print(sales_by_year[:, 1])
```

**Stock** **Data** **Example**

```python
import numpy as np

stock_returns = np.array([
    [2.1, 1.5, -0.8, 3.2, 1.7],
    [1.8, -1.2, 0.9, 2.8, 3.3],
    [0.5, 0.8, 1.2, -0.3, 2.1],
    [3.2, 2.8, 1.9, 0.7, -1.2],
    [1.1, 1.5, 2.0, 2.2, 1.8]
])

avg_returns = np.mean(stock_returns, axis=1)
print(avg_returns)

best_company = np.argmax(avg_returns)
print(best_company)

cumulative_returns = np.cumprod(1 + stock_returns/100, axis=1) * 100 - 100
print(cumulative_returns)
```
