```python
#else-Statement with For Loop
for num in range(10,20):                               #For loop to iterate between 10 to 20
  for i in range(2,num):                               #For loop to iterate on the factors
    if num%i == 0:                                     #If statement to determine the first factor
      j=num/i                                          #To calculate the second factor
      print('%d equals %d * %d' % (num,i,j))           #To move to the next number
      break
  else:
    print(num, 'is a prime number')
```

```python
# Defining a function to check if list items are positive or negative
def positive_or_negative():
   # Going through each item in the list
   for i in [5,6,7]:
      # Verifying if the current item is positive (greater than or equal to 0)
      if i >= 0:
         # If positive, display a message
         print("Positive number")
      else:
         # If negative, show message and exit the loop
         print("Negative number")
         break
   # This else runs only if the loop ends without hitting a break
   else:
      # Final message after successful loop completion
      print("Loop-else Executed")
# Executing the function
positive_or_negative()

```

```python
# Python Infinite while Loop(without false statement)

var = 1
while var == 1 : # This constructs an infinite loop
   num = int(input("Enter a number :"))
   print ("You entered: ", num)
print ("Good bye!")

'''You must be cautious when using while loops
because of the possibility that this condition never resolves to a FALSE value
'''
```

```python
# example-02 with false statement
var = 1
while var == 1:
   num = int(input("Enter a number: "))
   print("You entered:", num)
   if num == 0:   # If user enters 0
       var = 0    # Change var to 0 to stop the loop
print("Good bye!")
```

```python
# ......Break Statements....
# Break with with For-loop

for Letter in 'Tanisha':
  if Letter == 'h':
    break
  print('Letter: ' , Letter)
print("End of for loop")
```

```python
# break with while-loop
var = 10
while var > 0:
   print ('Current variable value :', var)
   var = var -1
   if var == 5:
      break

print ("Good bye!")
```

```python
#......Continue Statements.....
#Continue with For-Loop

for letter in 'Python':
   if letter == 'h':
      continue
   print ('Current Letter :', letter)
print ("Good bye!")
```

```python
#Continue with While-Loop

num = 60
print ("Prime factors for: ", num)
d=2
while num > 1:
   if num%d==0:
      print (d)
      num=num/d
      continue
   d=d+1
```

```python
#.....Pass Statement.....
# Pass with For-loop
for letter in 'Python':
   if letter == 'h':
      pass
      print ('This is pass block')
   print ('Current Letter :', letter)
print ("Good bye!")
```

```python
#.....Nested Loops.....
# Nested for-Loop
months = ["jan", "feb", "mar"]
days = ["sun", "mon", "tue"]

for x in months:
  for y in days:
    print(x, y)

print("Good bye!")
```
