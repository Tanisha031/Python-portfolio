Simple Class & Object

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

dog1 = Dog("Buddy", 3)

print(dog1.name)
print(dog1.age)
```

Class with Method

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

    def show(self):
        print("Car brand:", self.brand)

c1 = Car("Toyota")
c1.show()
```

Encapsulation (Private Variable)

```python
class Bank:
    def __init__(self):
        self.__balance = 1000

    def show_balance(self):
        print(self.__balance)

b = Bank()
b.show_balance()
```

Inheritance (Parent → Child)

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.speak()
d.bark()
```

Polymorphism (Same Method, Different Output)

```python
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

animals = [Dog(), Cat()]

for a in animals:
    a.sound()
```

Abstraction (Simple)

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Bark")

d = Dog()
d.sound()
```

Using super()

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)

d = Dog("Buddy")
print(d.name)
```

Simple str Method

```python
class Dog:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name

d = Dog("Buddy")
print(d)
```

Getter & Setter

```python
class Employee:
    def __init__(self):
        self.__salary = 50000

    def get_salary(self):
        return self.__salary

    def set_salary(self, s):
        if s > 0:
            self.__salary = s

e = Employee()
print(e.get_salary())

e.set_salary(60000)
print(e.get_salary())
```
