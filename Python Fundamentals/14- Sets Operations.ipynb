**PYTHON** - **JOINS** **SET**

UNION: Join set1 and set2 in one, remove duplicates

```python
set1 = {1,2,3,4,6,5}
set2 = {6,7,8,9,10, 5}
set3 = set1.union(set2)
print(set3)
```

use | this operator instead of union , you'll get the same result

```python
set1 = {1,2,3,4}
set2 = {'a', 'b', 'c','d'}
set3 = set1 | set2
print(set3)
```

join multiple sets using union method

```python
set1 = {1,2,3,4,5}
set2 = {'a', 'b', 'c', 'd'}
set3 = {'erum','sohail', 'iqra', 'afeera'}
set4 = {True, False, False}

set5 = set1.union(set2, set3, set4)
print(set5)
```

use | operator for multiple joins you'll get same results

```python
set1 = {1,2,3,4,5}
set2 = {'a', 'b', 'c', 'd'}
set3 = {'erum','sohail', 'iqra', 'afeera'}
set4 = {True, False, False}

set5 = set1 | set2 | set3 | set4
print(set5)
```

join set with tuple

```python
x = {'apple', 'banana', 'cherry'}
y = ('kiwi', 'orange')
z = x.union(y)
print(z)
```

UPDATE: this method insert all the item of set to current by taking from other sets. but it will change the existing/original set and doesnot return a new set. and remove duplicates

```python
set1 = {1,'z',3,'x',5}
set2 = {'z', 'y', '2', 'w', '3', 'b'}
set1.update(set2)
print(set1)
```

INTERSECTION:Keep ONLY the duplicates. The intersection() method will return a new set, that only contains the items that are present in both sets.

```python
set1 = {1,2,3,4,5}
set2 = {2,4,1,6,7}

set3 = set1.intersection(set2)
print(set3)
```

you can also use & operator for this

```python
set1 = {1,2,3,4,5}
set2 = {2,4,1,6,7}

set3 = set1 & set2
print(set3)
```

intersection_update: will keep only duplicate values. but it will change the original set instead of creating new set  

```python
set1 = {'facebook', 'snapchat', 'instagram', 'kisskh'}
set2 = {'snapchat', 'easypaise', 'telenor', 'gmail', 'facebook'}

set3 = set1.intersection_update(set2)
print(set3)
```

join sets, which contain True, 1, False 0. and see what it's considered

```python
set1 = {'facebook', 1, 'instagram', 0, 'cherry'}
set2 = {False, 'easypaise', True, 'gmail', 'facebook'}

set3 = set1.intersection(set2)
print(set3)
```

Difference: method will  return new set that will contain only item in first set that are not present in second set

```python
set1 = {'apple', 'banana', 'cherry', 'grapes', 'mango'}
set2 = {'mango', 'pineapple', 'cheekoo', 'apple'}

set3 = set1.difference(set2)
print(set3)
```

you can also use minus (-) for difference

```python
set1 = {'apple', 'banana', 'cherry', 'mango'}
set2 = {'mango', 'banana', 'grapes', 'pineapple'}

set3 = set1 - set2
print(set3)
```

difference_update:keep the unique item from first set that are not present in second. change the original instead of create new set

```python
set1 = {'apple','banana', 'cherry', 'grapes', 'mango'}
set2 = {'mango', 'pineapple', 'cheekoo', 'apple'}

set3 = set1.difference_update(set2)
print(set3)
```

systemetric_difference: only keep those values which are not present in both

```python
set1 = {'apple', 'banana', 'cherry', 'grapes', 'mango'}
set2 = {'mango', 'pineapple', 'cheekuu', 'orange'}

set3 = set1.symmetric_difference(set2)
print(set3)
```

you can use ^ instead of symmetic_method, you'll get same result

```python
set1 = {'apple', 'banana', 'cherry', 'mango', 'grapes'}
set2 = {'mango', 'pineapple', 'mango', 'cheekuu'}

set3 = set1 ^ set2
print(set3)
```

symmetric_difference_update() method will also keep all but the duplicates, but it will change the original set instead of returning a new set.

```python
set1 = {'apple', 'banana', 'cherry', 'grapes', 'mango'}
set2 = {'mango', 'pineapple', 'cheekoo', 'apple'}

set1.symmetric_difference_update(set2)
print(set1)
```
