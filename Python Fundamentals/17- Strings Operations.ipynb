Check String : if a certain phrase or chahrater presenet in string.
using keyword 'in'.

```python
# without if condition
text = 'Tanisha Maheshwari, 22 years old'
print('22' in text)
```

```python
# with string condition
txt = "Hello myself tanisha mahehswari and i'm junior data analyst"
if 'junior' in txt:
  print("yes 'junior' is present")
else:
  print("try another")
```

Check if not: checks if certain phrase or character is not present in string. using keyword 'not in'.

```python
# without if condition
txt = "Hello myself tanisha mahehswari and i'm junior data analyst having 6 month experience in this field"
print('fresh graduate' not in txt)
```

```python
# with if condition
txt = "Hello myself tanisha mahehswari and i'm junior data analyst having 6 month experience in this field"
if 'fresh graduate' not in txt:
  print("yes 'fresh graduate' is not present")
```

**PYTHON** - **String** **Format**

---




format the string by puting simply f front of string literal and add {} brcaets for placeholder for a vaiable or other operation.

```python
age = 22
txt = f"myself tanisha and i'm {age} year old"
print(txt)
```

placeholder can contain variable, operations, functions, and add modifier

```python
price = 59
txt = f"This shirt prize is {price} dollar."
print(txt)
```

A modifier is included by adding a colon : followed by a legal formatting type, like .2f which means fixed point number with 2 decimals

```python
price = 100
txt = f"this painting prize is {price:.2f} dollar."
print(txt)
```

Placeholder can also contain math operation

```python
txt = f'the price of this cosmetic box is {12*25} dollar'
print(txt)
```

**PYTHON** - **Escape** **character**

To insert illegal characters in a string, use a backslash \ followed by the character escape character

```python
# using double quotes inside a double quotes can be illegal and might get erro.
# that's why use backslash for single and double quotes.

txt = "we are the so-called \"Vikings\" from the north."
print(txt)
```

Below are other method for Escape character

```python
# Single Quote
txt1 = 'It\'s a beautiful day.'

# Backslash
txt2 = "This is a backslash: \\"

# New Line
txt3 = "First line\nSecond line"

# Carriage Return
txt4 = "Hello\rWorld"

# Tab
txt5 = "Name:\tJohn"

# Backspace
txt6 = "Hello\bWorld"

# Form Feed
txt7 = "Page 1\fPage 2"

# Octal value (\ooo)
txt8 = "Octal: \101"  # 'A'

# Hex value (\xhh)
txt9 = "Hex: \x41"    # 'A'

print('-', txt1)
print('-',txt2)
print('-',txt3)
print('--',txt4)
print('-',txt5)
print('-',txt6)
print('-',txt7)
print('-',txt8)
print('-',txt9)
```
