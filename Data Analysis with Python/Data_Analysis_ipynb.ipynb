### **Data Analysis with Numpy**

**1. NumPy — Array Basics**

```python
# Import NumPy
import numpy as np

# Create a 2D array
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Display array details
print("Array:\n", arr)
print("Shape:", arr.shape)     # Rows and columns
print("Dimensions:", arr.ndim) # Number of dimensions
```

**2. Creating NumPy Arrays**

```python
import numpy as np

# Different ways to create arrays
a = np.array([1, 2, 3])        # From list
b = np.zeros((2, 2))           # Matrix of zeros
c = np.ones((2, 2))            # Matrix of ones
d = np.arange(0, 10, 2)        # Range with step
e = np.empty((2, 2), dtype=int) # Empty array

print("Matrix A \n", a)
print("Matrix B \n", b)
print("Matrix C \n", c)
print("Matrix D \n", d)
print("Matrix E \n", e)
```

**3. NumPy Array Operations**

```python
import numpy as np

a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

# Element-wise operations
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

**4. NumPy Indexing**

```python
import numpy as np

arr = np.array([10, 20, 30, 40])

# Access elements
print("First element:", arr[0])
print("Last element:", arr[-1])
```

**5. NumPy Slicing**

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

# Slicing examples
print("Slice (1 to 3):", arr[1:4])
print("First three elements:", arr[:3])
```

**6. NumPy Broadcasting**

```python
import numpy as np

arr = np.array([1, 2, 3])

# Scalar broadcasting
print("Add scalar:", arr + 5)

# Array broadcasting
b = np.array([[1], [2], [3]])
c = np.array([10, 20, 30])

print("Broadcasting result:\n", b + c)
```

**9. Pandas CRUD Operations**

```python
import pandas as pd

# Create DataFrame
data = {"Name": ["A", "B", "C"], "Age": [20, 21, 22]}
df = pd.DataFrame(data)

# Create (Save to CSV)
df.to_csv("data.csv", index=False)

# Read
df = pd.read_csv("data.csv")
print("Read:\n", df)

# Update
df["Age"] = df["Age"] + 1
df["City"] = ["Delhi", "Mumbai", "Pune"]
print("\nUpdate:\n", df)

# Delete
df = df.drop("City", axis=1)  # Drop column
df = df.drop(0, axis=0)       # Drop row
print("\nDelete:\n", df)
```

### **Exploratory Data Analysis (EDA)**

**10. Data Inspection**

```python
import pandas as pd

data = {
    "Name": ["A", "B", "C", "A", "B"],
    "Age": [20, 21, 22, 20, 21]
}

df = pd.DataFrame(data)

# Dataset overview
print(df.info())

# Statistical summary
print(df.describe())

# Value counts
print(df["Name"].value_counts())

# Preview data
print(df.head())
print(df.tail())
```

**11. Indexing & Selection**

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Age": [20, 21, 22]
})

# Column selection
print(df["Name"])

# Row slicing
print(df.iloc[0:2])
```

**12. Grouping & Aggregation**

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B", "A"],
    "Marks": [80, 90, 85]
})

# Group by and calculate mean
print(df.groupby("Name")["Marks"].mean())
```

**13. Merging DataFrames**

```python
import pandas as pd

df1 = pd.DataFrame({"Name": ["A", "B"], "Age": [20, 21]})
df2 = pd.DataFrame({"Name": ["A", "B"], "City": ["Delhi", "Mumbai"]})

# Merge on common column
print(pd.merge(df1, df2, on="Name"))
```

**14. Sorting**

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Marks": [85, 95, 90]
})

# Sort values
print(df.sort_values(by="Marks", ascending=False))
```

**15. Filtering**

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Marks": [85, 95, 90]
})

# Filter rows
print(df[df["Marks"] > 88])
```

**16. Index Handling**

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Marks": [85, 95, 90]
})

# Set index
df_indexed = df.set_index("Name")
print(df_indexed)

# Reset index
print(df_indexed.reset_index())
```

**17. Missing Data Handling**

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Age": [20, None, 22]
})

# Check missing values
print(df.isnull())
print(df.isnull().sum())

# Drop missing values
print(df.dropna())

# Fill missing values
df["Age"].fillna(df["Age"].mean(), inplace=True)
print(df)
```

**18. Duplicate Handling**

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B", "A", "C"],
    "Age": [20, 21, 20, 22]
})

# Check duplicates
print(df.duplicated())
print(df.duplicated().sum())

# Remove duplicates
df_clean = df.drop_duplicates()
print(df_clean)
```

**19. Outlier Detection (IQR)**

```python
import pandas as pd

df = pd.DataFrame({
    "Values": [10, 12, 14, 15, 100]
})

Q1 = df["Values"].quantile(0.25)
Q3 = df["Values"].quantile(0.75)
IQR = Q3 - Q1

# Detect outliers
outliers = df[(df["Values"] < Q1 - 1.5*IQR) | (df["Values"] > Q3 + 1.5*IQR)]
print(outliers)
```

**20. Outlier Detection (Z-Score)**

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Values": [10, 12, 14, 15, 100]
})

mean = np.mean(df["Values"])
std = np.std(df["Values"])

df["Z_score"] = (df["Values"] - mean) / std

# Detect outliers
outliers = df[df["Z_score"].abs() > 3]
print(outliers)
```

### **Data Visualization — Matplotlib**

**21. Line Plot**

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axis([0, 6, 0, 20])
plt.show()
```

**22. Bar Chart**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["target"] = iris.target

counts = df["target"].value_counts()

plt.bar(counts.index, counts.values)
plt.title("Bar Chart")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

**23. Histogram**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

plt.hist(df["sepal length (cm)"], bins=10)
plt.title("Histogram")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

**24. Scatter Plot**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

plt.scatter(df["sepal length (cm)"], df["sepal width (cm)"])
plt.title("Scatter Plot")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

**25. Box Plot**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

plt.boxplot(df["sepal width (cm)"], labels=["Sepal Width"])
plt.title("Box Plot")
plt.ylabel("Value")
plt.show()
```

**26. Correlation Heatmap**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

corr = df.corr()

plt.imshow(corr, cmap='autumn', interpolation='nearest')
plt.title("Correlation Heatmap")
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.show()
```

### **Seaborn Visualization**

**27. Seaborn Scatter Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target

# Seaborn scatter plot with hue
sns.scatterplot(
    x="sepal length (cm)",
    y="sepal width (cm)",
    hue="species",
    data=df
)

plt.title("Seaborn Scatter Plot")
plt.show()
```
