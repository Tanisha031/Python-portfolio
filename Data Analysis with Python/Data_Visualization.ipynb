### **Visualization with Matplotlib**

**1a- Scatter Plot**

```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the database
data = pd.read_csv("tips.csv")

# Scatter plot with day against tip
plt.scatter(data['day'], data['tip'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

### **Visualization with Bokeh**

**1**- **Scatter** **Plot**

```python
from bokeh.plotting import figure, output_file, show
from bokeh.palettes import magma
import pandas as pd

# create figure
graph = figure(title="Bokeh Scatter Graph")

# reading the database
data = pd.read_csv("tips.csv")

color = magma(256)

# plotting
graph.scatter(data['total_bill'], data['tip'], color=color)

# show plot
show(graph)
```

**2- Line Chart**

```python
from bokeh.plotting import figure, output_file, show
import pandas as pd

# create figure
graph = figure(title="Bokeh Bar Chart")

# reading the database
data = pd.read_csv("tips.csv")

# count values
df = data['tip'].value_counts()

# plotting
graph.line(df, data['tip'])

# show plot
show(graph)
```

**3a**- **Bar** **Chart**

```python
from bokeh.plotting import figure, output_file, show
import pandas as pd

# create figure
graph = figure(title="Bokeh Bar Chart")

# reading the database
data = pd.read_csv("tips.csv")

# plotting
graph.vbar(data['total_bill'], top=data['tip'])

# show plot
show(graph)
```

**3b- Bar Chart Interactive Legend**

```python
from bokeh.plotting import figure, output_file, show
import pandas as pd

# create figure
graph = figure(title="Bokeh Bar Chart")

# reading the database
data = pd.read_csv("tips.csv")

# plotting bars
graph.vbar(data['total_bill'], top=data['tip'],
           legend_label="Bill VS Tips", color='green')

graph.vbar(data['tip'], top=data['size'],
           legend_label="Tips VS Size", color='red')

# interactive legend
graph.legend.click_policy = "hide"

# show plot
show(graph)
```

**4- Widgets (Button, Checkbox, Radio)**

```python
from bokeh.io import show
from bokeh.models import Button, CheckboxGroup, RadioGroup, CustomJS

# Button
button = Button(label="GFG")
button.js_on_click(CustomJS(
    code="console.log('button: click!', this.toString())"))

# labels
L = ["First", "Second", "Third"]

# Checkbox
checkbox_group = CheckboxGroup(labels=L, active=[0, 2])
checkbox_group.js_on_click(CustomJS(code="""
    console.log('checkbox_group: active=' + this.active, this.toString())
"""))

# Radio
radio_group = RadioGroup(labels=L, active=1)
radio_group.js_on_click(CustomJS(code="""
    console.log('radio_group: active=' + this.active, this.toString())
"""))

show(button)
show(checkbox_group)
show(radio_group)
```

**5- Slider**

```python
from bokeh.io import show
from bokeh.models import CustomJS, Slider

# Slider
slider = Slider(start=1, end=20, value=1, step=2, title="Slider")

slider.js_on_change("value", CustomJS(code="""
    console.log('slider: value=' + this.value, this.toString())
"""))

show(slider)
```

**5- Dropdown Menu**

```python
import plotly.graph_objects as px
import pandas as pd

# reading the database
data = pd.read_csv("tips.csv")

plot = px.Figure(data=[px.Scatter(
    x=data['day'],
    y=data['tip'],
    mode='markers')
])

# dropdown menu
plot.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(args=["type", "scatter"], label="Scatter Plot", method="restyle"),
                dict(args=["type", "bar"], label="Bar Chart", method="restyle")
            ]),
            direction="down",
        ),
    ]
)

plot.show()
```

**6- Buttons**

```python
import plotly.graph_objects as px
import pandas as pd

# reading the database
data = pd.read_csv("tips.csv")

plot = px.Figure(data=[px.Scatter(
    x=data['day'],
    y=data['tip'],
    mode='markers')
])

# buttons
plot.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            direction="left",
            buttons=list([
                dict(args=["type", "scatter"], label="Scatter Plot", method="restyle"),
                dict(args=["type", "bar"], label="Bar Chart", method="restyle")
            ]),
        ),
    ]
)

plot.show()
```

**7- Slider & Range Selector**

```python
import plotly.graph_objects as px
import pandas as pd

# reading the database
data = pd.read_csv("tips.csv")

plot = px.Figure(data=[px.Scatter(
    y=data['tip'],
    mode='lines')
])

# slider and selector
plot.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, step="day", stepmode="backward"),
            ])
        ),
        rangeslider=dict(visible=True),
    )
)

plot.show()
```
