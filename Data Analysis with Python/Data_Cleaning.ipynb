**Import Required Libraries**

```python
# Import pandas for data manipulation and analysis
import pandas as pd

# Import numpy for numerical operations
import numpy as np
```

**Load** **a** **Dataset**

```python
# URL of Titanic dataset (public dataset from GitHub)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Read the dataset into a pandas DataFrame
df = pd.read_csv(url)
```

**Understand the Dataset**

```python
# Display first 5 rows to understand data structure
df.head()
```

```python
# Show summary of dataset (columns, data types, missing values)
df.info()
```

```python
# Show statistical summary of numeric columns
df.describe()
```

**Check** **Missing** **Values**

```python
# Count missing values in each column
df.isnull().sum()
```

**Handle** **Missing** **Data**

```python
# Fill missing values in Age column with mean value of Age
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
# Fill missing values in Embarked column with most frequent value (mode)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

**Remove** **Duplicates**

```python
# Count duplicate rows in dataset
df.duplicated().sum()
```

```python
# Remove duplicate rows from dataset
df.drop_duplicates(inplace=True)
```

**Fix** **Data** **Types**

```python
# Check data types of all columns
df.dtypes
```

```python
# Convert Survived column to integer type (if needed)
df['Survived'] = df['Survived'].astype(int)
```

**Handle** **Outliers (IQR Method)**

```python
# Calculate first quartile (25th percentile)
Q1 = df['Fare'].quantile(0.25)

# Calculate third quartile (75th percentile)
Q3 = df['Fare'].quantile(0.75)

# Calculate Interquartile Range (IQR)
IQR = Q3 - Q1

# Define lower boundary for outliers
lower_bound = Q1 - 1.5 * IQR

# Define upper boundary for outliers
upper_bound = Q3 + 1.5 * IQR

# Keep only rows within acceptable range (remove outliers)
df = df[(df['Fare'] >= lower_bound) & (df['Fare'] <= upper_bound)]
```

**Rename Columns**

```python
# Rename column for better readability
df.rename(columns={'PassengerId': 'Passenger_ID'}, inplace=True)
```

**Remove** **Unnecessary** **Columns**

```python
# Remove Cabin column (too many missing values, not useful here)
df.drop(['Cabin'], axis=1, inplace=True)
```

**Clean Text Data**

```python
# Convert all values in Sex column to lowercase for consistency
df['Sex'] = df['Sex'].str.lower()
```

**Final** **Clean** **Dataset** **Check**

```python
# Show first 5 rows of cleaned dataset
df.head()
```
