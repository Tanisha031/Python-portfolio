# **Color** **Palette**

 **Default/Qualitative/(Categorical) Color** **Palette**:
Used when you have categories (no order).

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# current colot palette
palette = sns.color_palette()

# plots the color palette as a
# horizontal array
sns.palplot(palette)

plt.show()
```

**Setting the default Color Palette**

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")
def plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# setting the default color palette
sns.set_context(context='paper', font_scale=1)
sns.set_palette('vlag')
plt.subplot(211)
plot()

# setting another default color palette
sns.set_context(context= 'paper', font_scale=1)
sns.set_palette('Accent')
plt.subplot(212)
plot()

plt.show()
```

```python
# BarPlot

import seaborn as sns
import matplotlib.pyplot as plt

# Set a Seaborn color palette
sns.set_palette("Set2")

# Create some example data
data = [3, 5, 7, 9, 2, 4, 6, 8]

# Create a simple bar plot using the set color palette
sns.barplot(x=range(len(data)), y=data)

plt.show()
```
