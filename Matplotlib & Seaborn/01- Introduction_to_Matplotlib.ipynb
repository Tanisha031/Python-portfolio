```python
# basic plot, and understand basic components of matplotlib figure.
x = [0,2,4,6,8]
y = [0,4,16,36,64]

fig, ax = plt.subplots()
ax.plot(x, y, marker='o', label="Data Points")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_title("Simple Line Plot")
plt.show()

# ax.plot(x,y)
# plt.show()
```
