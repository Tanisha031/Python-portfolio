**Some basic plots using seaborn**

```python
import numpy as np
import seaborn as sns

# Histplot:
sns.set_style('white')
x = np.random.RandomState(10)
d = x.normal(size = 100)
sns.histplot(d, kde = True, color = 'm')
```

```python
# Distplot

sns.set(style="white")

# Generate a random univariate dataset
rs = np.random.RandomState(10)
d = rs.normal(size=100)

# Define the colors to use
colors = ["r", "g", "b"]

# Plot a histogram with multiple colors
sns.distplot(d, kde=True, hist=True, bins=10,
             rug=True,hist_kws={"alpha": 0.3,
                                "color": colors[0]},
             kde_kws={"color": colors[1], "lw": 2},
             rug_kws={"color": colors[2]})
```

```python
# Lineplot
import seaborn as sns

sns.set(style="dark")
fmri = sns.load_dataset("fmri")

# Plot the responses for different\
# events and regions
sns.lineplot(x="timepoint", y="signal", hue="region", style="event", data=fmri)
```

```python
# Lmplot:

import seaborn as sns
sns.set(style="ticks")

# Loading the dataset
df = sns.load_dataset("anscombe")

# Show the results of a linear regression
sns.lmplot(x="x", y="y", data=df)
```

**Plotting graph using Seaborn**

```python
# Stripplot - Plotting categorical scatter

# importing the required module
import matplotlib.pyplot as plt
import seaborn as sns

# x axis values
x =['sun', 'mon', 'fri', 'sat', 'tue', 'wed', 'thu']
# y axis values
y =[5, 6.7, 4, 6, 2, 4.9, 1.8]
xy = [x, y]

# plotting strip plot with seaborn
ax = sns.stripplot(xy);
# giving labels to x-axis and y-axis
ax.set(xlabel ='Days', ylabel ='Amount_spend')

# giving title to the plot
plt.title('My first graph');
# function to show plot
plt.show()
```

```python
#Stripplot using inbuilt data-set

import matplotlib.pyplot as plt
import seaborn as sns

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
iris = sns.load_dataset('iris')

# plotting strip plot with seaborn
ax = sns.stripplot(x='species', y='sepal_length', data=iris)

# giving title to the plot
plt.title('Graph')

# function to show plot
plt.show()
```

```python
# Swarmplot

# importing the required module
import matplotlib.pyplot as plt
import seaborn as sns

# use to set style of background of plot
sns.set(style="whitegrid")
# loading data-set
iris = sns.load_dataset('iris')

# plotting swarmplot plot with seaborn
ax = sns.swarmplot(x='species', y='sepal_length', data=iris)

# giving title to the plot
plt.title('Graph')
# function to show plot
plt.show()
```

```python
# change the representation of data on a particular axis

# importing the required module
import matplotlib.pyplot as plt
import seaborn as sns
# use to set style of background of plot
sns.set(style="whitegrid")
# loading data-set
iris = sns.load_dataset('iris')

# plotting swarmpplot with seaborn
ax = sns.swarmplot(x='sepal_length', y='species', data=iris)

# giving title to the plot
plt.title('Graph')
# function to show plot
plt.show()
```

```python
# Barplot

# import the seaborn library
import seaborn as sns
# reading the dataset
df = sns.load_dataset('tips')

# change the estimator from mean to
# standard deviation
sns.barplot(x ='sex', y ='total_bill', data = df,  palette ='plasma')
```

```python
# Countplot
# import the seaborn library
import seaborn as sns

# reading the dataset
df = sns.load_dataset('tips')

sns.countplot(x ='sex', data = df)
```

```python
# Boxplot
# import the seaborn library
import seaborn as sns

# reading the dataset
df = sns.load_dataset('tips')

sns.boxplot(x='day', y='total_bill', data=df, hue='smoker')
```

```python
# Violinplot
# import the seaborn library
import seaborn as sns

# reading the dataset
df = sns.load_dataset('tips')
sns.violinplot(x='day', y='total_bill', data=df,
               hue='sex', split=True)
```
