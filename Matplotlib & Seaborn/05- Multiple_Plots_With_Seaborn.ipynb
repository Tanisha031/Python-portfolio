# **Multiple plots with Seaborn**

**Using Matplotlib** : add_axes(), subplot(), and subplot2grid().

```python
# Example - 01 : add_axes()

# importing packages
import seaborn as sns
import matplotlib.pyplot as plt


# loading dataset
data = sns.load_dataset("iris")
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Creating a new figure with width = 5 inches
# and height = 4 inches
fig = plt.figure(figsize =(5, 4))

# Creating first axes for the figure and plot the graph
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
graph()

# Creating second axes for the figure and plot the graph
ax2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])
graph()

plt.show()
```

```python
# Example - 2: Usingsubplot() method

# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Adding the subplot at the specified grid position
plt.subplot(121)
graph()

# Adding the subplot at the specified grid position
plt.subplot(122)
graph()

plt.show()
```

```python
# Example 3: Using subplot2grid() method

# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# adding the subplots
axes1 = plt.subplot2grid (
  (7, 1), (0, 0), rowspan = 2,  colspan = 1)
graph()

axes2 = plt.subplot2grid (
  (7, 1), (2, 0), rowspan = 2, colspan = 1)
graph()

axes3 = plt.subplot2grid (
  (7, 1), (4, 0), rowspan = 2, colspan = 1)
graph()
```

**Using** **Seaborn** : FacetGrid(),  PairGrid()

```python
# Method 1: Using FacetGrid() method

# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

plot = sns.FacetGrid(data, col="species")
plot.map(plt.plot, "sepal_width")

plt.show()
```

```python
# Method 2: Using PairGrid() method

# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("flights")

plot = sns.PairGrid(data)
plot.map(plt.plot)

plt.show()
```
